```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from IPython.display import clear_output

from Grid1D import TF_Grid_1D
from Data import DataFile
```

# Mutual information of $h_{tf}$ and occupancy

## Occupancies for a range of $h_{tf}$  values

```python
from Experiments import average_occupancy_plot

eb = 10.5
J  = 9
htf_list = np.linspace(-17,-13,17)

N = 6
padding = 10
L = N+2*padding
sites = np.arange(N)+10

T = 22500000
runs = 500

occupancies = average_occupancy_plot(runs, htf_list, (eb,T), 'htf', '$h_{tf}$', TF_Grid_1D, \
                                     NSites=len(sites), L=L, J=J, sites=sites, predict=False)

plt.xlim([htf_list[0],htf_list[-1]])
plt.ylim([-0.1,1.1])
plt.yticks(np.arange(6)*0.2)
plt.title('$e_b$ = '+f'{eb}'+', $J$ = '+f'{J}')
plt.legend('',frameon=False)
```

## Distribution of occupancies given $h_{tf}$ ($p(occupancy|h_{tf})$)

*Continuous (histogram) distribution:*

```python
N_bins = 50
p_occupation_given_htf = np.zeros((len(htf_list),N_bins))

for i,htf in enumerate(htf_list):
    # Plot histogram
    plt.figure()
    hist_values, bin_edges, _ = plt.hist(occupancies[i], N_bins, [0,1], density=True)
    plt.xlim([0,1])
    plt.title('$h_{tf}$ = '+f'{htf}')
    
    # Store distribution
    p_occupation_given_htf[i,:] = hist_values*(1/N_bins) # Discretized distribution

print(np.sum(p_occupation_given_htf, axis=1))
#assert(np.all(np.sum(p_occupation_given_htf, axis=1)==1))
```

*Fitting a gaussian distribution:*

```python
import math as m

gaussian = lambda x, mu, sigma: (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(1/2)*((x-mu)/(sigma))**2)

def compact_gaussian(x, mu, sigma):
    gaussian_01 = gaussian(x, mu, sigma)
    gaussian_01[0] += (1+m.erf(mu/(np.sqrt(2)*sigma)))/2
    gaussian_01[1] += m.erfc((1-mu)/(np.sqrt(2)*sigma))/2
    return gaussian_01
```

```python
N_bins = 50
p_occupation_given_htf = np.zeros((len(htf_list),N_bins))

for i,htf in enumerate(htf_list):
    # Plot histogram
    plt.figure()
    hist_values, bin_edges, _ = plt.hist(occupancies[i], N_bins, [0,1], density=True)
    plt.xlim([0,1])
    plt.title('$h_{tf}$ = '+f'{htf}')
    
    # Fit distribution
    x = bin_edges[:-1]+1/(2*N_bins)
    vars_opt, _ = sp.optimize.curve_fit(compact_gaussian, x, hist_values, bounds = ([0,0], [1,10]))
    plt.plot(x, compact_gaussian(x, vars_opt[0], vars_opt[1]))
    
    # Store distribution
    p_occupation_given_htf[i,:] = compact_gaussian(x, vars_opt[0], vars_opt[1])*(1/N_bins) # Discretized distribution
    
```

*Fitting a beta distribution:*

```python
def beta_distribution(x, a, b):
    return sp.stats.beta.pdf(x, a, b)
```

```python
N_bins = 50
p_occupation_given_htf = np.zeros((len(htf_list),N_bins))

for i,htf in enumerate(htf_list):
    # Plot histogram
    plt.figure()
    hist_values, bin_edges, _ = plt.hist(occupancies[i], N_bins, [0,1], density=True)
    plt.xlim([0,1])
    plt.title('$h_{tf}$ = '+f'{htf}')
    
    # Fit distribution
    x = bin_edges[:-1]+1/(2*N_bins)
    vars_opt, _ = sp.optimize.curve_fit(beta_distribution, x, hist_values, bounds = ([0,0], [10,100]))
    plt.plot(x, beta_distribution(x, vars_opt[0], vars_opt[1]))
    
    # Store distribution
    p_occupation_given_htf[i,:] = beta_distribution(x, vars_opt[0], vars_opt[1])*(1/N_bins) # Discretized distribution
    
```

Some of the fits are much better than the gaussian, but I think the ability of the beta distribution to fit the histogram is limited due to the bounds set for fitting.

## Mutual infromation, $I$

```python
N = len(occupancies)
p_htf = 1/N
p_occupation_and_htf = p_occupation_given_htf*p_htf
p_occupation = np.sum(p_occupation_and_htf, axis=0) # axis 0 is htf values, axis 1 is probailities

nonzero_terms = np.where(p_occupation_and_htf!=0)
I = np.sum((p_occupation_and_htf*np.log2(p_occupation_and_htf/(p_occupation*p_htf)))[nonzero_terms])
print('Mutual information estimate:',I)
```

```python
print(np.log2(17)) # 17 different htf values
```

- For J = 0   (50 bins) Mutual information estimate: 1.2557695628279695 (occupancy remains extremely low <0.2)
- For J = 4.5 (50 bins) Mutual information estimate: 1.3873791037418983
- For J = 9   (50 bins) Mutual information estimate: 1.0549331012317182 (very large standard deviation)

For very low $J$, there is little noise, but the range of occupancies obtained for the realistic $h_{tf}$ range is very small. Therefore, the information is low. If $J$ increases, both the range of occupancies and the standard deviation increase. Initially this leads to a small increase in the mutual information. However, when the occupancy range can not grow much bigger, but the standard deviation keeps increasing, the mutual information decreases again. There is thus an optimal value of $J$ for which the mutual inforation is highest. This optimal value of $J$ will likely depend on $e_b$ and on the averaging time $T$.
