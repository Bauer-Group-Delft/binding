```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

# Experiment 1: Comparing analytical standard deviation with simulation results

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

beta  = 1
htf   = -15
eb    = 11    
J     = 4
L     = 4
sites = np.arange(L)

T0 = 45000000
multiples = np.logspace(0, 2)
T_list = [m*T0 for m in multiples]

mode = 'standard'
periodic = True

R = FullRates(mode, htf=htf, eb=eb, J=J)
E = FullEnergies(mode, htf, eb, J)
M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
mat = M.create_Matrix(sparse=False)
occupancies = M.states.calculate_state_occupancies(sites)
P_steady = M.calculate_steady_state()
```

*Simulation starting from steady state:*

```python
from Data import DataFile
from Gillespie import GillespieSim
from Grid1D import TF_Grid_1D
from Tests import time_mean

def steady_state_time_average_of_occupancy(L, htf, eb, J, sites, T_list, runs, str_states, P_steady):
    
    SimulationGrid = TF_Grid_1D(L, htf, eb, J, sites)
    datafile = DataFile(None)
    Simulation = GillespieSim(SimulationGrid, datafile)

    mean_occupancy_list = []
    std_occupancy_list  = []

    for T in T_list:
        time_means = []
        for run in range(runs):
            print('run', run)

            # Randomly pick a grid based on the steady state probabilities
            str_grid = np.random.choice(str_states, p=P_steady)
            grid = np.array([int(s) for s in str_grid])

            # Update the grid to match the chosen grid
            Simulation.reinitialize(grid)
            
            # Run for a time T
            Simulation.run(T)

            # Calculate and store the mean occupancy
            times, occupancy_list = datafile.get_NBoundPerSite()
            time_means.append(time_mean(times, occupancy_list))

            clear_output(wait=True)

        # Calculate the mean and standard deviation over the runs
        mean_occupancy_list.append(np.mean(time_means))
        std_occupancy_list.append(np.std(time_means))
        
    return mean_occupancy_list, std_occupancy_list
```

```python
runs = 100

mean_occupancy_list, std_occupancy_list = steady_state_time_average_of_occupancy(L, htf, eb, J, sites, T_list, runs, M.str_states, P_steady)
```

*Analytcal limit equations:*

```python
def matrix_exponential_integral_limit(mat, T_inf=0):
    eigenvalues, eigenvectors = np.linalg.eig(mat)
    order = np.argsort(np.abs(eigenvalues))
    P = eigenvectors[:,order]

    D_int = np.zeros((len(eigenvalues),len(eigenvalues)))
    D_int[0,0] = T_inf
    D_int[1:,1:] = np.diag(-1/(eigenvalues[order])[1:])

    integral = P@D_int@np.linalg.inv(P)
    
    return integral
```

```python
integral = matrix_exponential_integral_limit(mat, 0)

correlation_integral_occupancy = occupancies@integral@(occupancies*P_steady)
std_occupancy_list_limit = [np.sqrt(2/T*correlation_integral_occupancy) for T in T_list]

mu_occupancy = occupancies@P_steady
sigma_squared_occupancy = (occupancies-mu_occupancy)**2@P_steady
tau_occupancy = 1/sigma_squared_occupancy*correlation_integral_occupancy
```

*Analytical non-limit equations:*

```python
eigenvalues, eigenvectors = np.linalg.eig(mat)   
order = np.argsort(np.abs(eigenvalues))
eigs = eigenvalues[order]
P = eigenvectors[:,order]

std_occupancy_list_analytical  = []

for T in T_list:
    D_int = np.zeros((len(eigs),len(eigs)))
    D_int[1:,1:] = np.diag((-2/(eigs[1:]*T)**2)*(1-np.exp(eigs[1:]*T)+eigs[1:]*T))
    integral = P@D_int@np.linalg.inv(P)
    
    var_occupancy = (occupancies@integral@(occupancies*P_steady))
    std_occupancy_list_analytical.append(np.sqrt(var_occupancy))
```

*Comparison:*

```python
xs = 10*np.array(multiples)

## Linear plot

plt.figure()
plt.plot(xs, std_occupancy_list, label='simulation result')
plt.plot(xs, std_occupancy_list_limit, label='analytical prediction ($T\gg\\tau_n$)')
plt.plot(xs, std_occupancy_list_analytical, label='analytcial prediction')
plt.plot([10*tau_occupancy/T0, 10*tau_occupancy/T0], [0,1], 'k:', label='correlation time, $\\tau$')
plt.xlabel('Averaging time $T$ in min.')
plt.ylabel('Standard deviation of mean occupancy')
plt.legend()

## Logarithic plot

plt.figure()
plt.loglog(xs, std_occupancy_list, label='simulation result')
plt.loglog(xs, std_occupancy_list_limit, label='analytical prediction ($T\gg\\tau_n$)')
plt.loglog(xs, std_occupancy_list_analytical, label='analytcial prediction')
plt.ylabel('Standard deviation of mean occupancy')
plt.legend()
```

When $T>>\tau$, the analytical prediction for teh limit case predicts the simulation results well. The simulations are initialized with randomly chosen grids based on the steady state probabilities, such that it is not necessary to equilibrate the system. The predictions from the analytical model only work for steady state (so far).

$log(a^b)=b\cdot log(a)$

We see that the predicted standard deviation in the limit ($T>>\tau_n$) scales as $1/sqrt(T)$ as expected. However, for the simulation results, the standard deviation initially decreases slower than the prediction.

We can estimate the standard deviation more accurately if we use the equations from Kaizu before the limit $T>>\tau$ is taken.


# Experiment 2: Dependence of standard deviation on parameter values

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

def calculate_mean_and_std_limit(htf, eb, J, L, sites, T, mode = 'standard', periodic = True):
    
    R = FullRates(mode, htf=htf, eb=eb, J=J)
    E = FullEnergies(mode, htf, eb, J)
    M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
    mat = M.create_Matrix(sparse=False)
    occupancies = M.states.calculate_state_occupancies(sites)
    P_steady = M.calculate_steady_state()

    mean_occupancy = occupancies@P_steady
    
    integral = matrix_exponential_integral_limit(mat, 0)
    std_occupancy = np.sqrt(2/T*(occupancies@integral@(occupancies*P_steady)))
    
    return mean_occupancy, std_occupancy
```

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

def calculate_mean_and_std(htf, eb, J, L, sites, T, mode = 'standard', periodic = True):
    
    # Set up matrix
    R = FullRates(mode, htf=htf, eb=eb, J=J)
    E = FullEnergies(mode, htf, eb, J)
    M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
    mat = M.create_Matrix(sparse=False)
    occupancies = M.states.calculate_state_occupancies(sites)
    P_steady = M.calculate_steady_state()

    # Mean occupancy
    mean_occupancy = occupancies@P_steady
    
    # Std occupancy
    eigenvalues, eigenvectors = np.linalg.eig(mat)   
    order = np.argsort(np.abs(eigenvalues))
    eigs = eigenvalues[order]
    P = eigenvectors[:,order]
    
    D_int = np.zeros((len(eigs),len(eigs)))
    D_int[1:,1:] = np.diag((-2/(eigs[1:]*T)**2)*(1-np.exp(eigs[1:]*T)+eigs[1:]*T))
    integral = P@D_int@np.linalg.inv(P)

    std_occupancy = np.sqrt((occupancies@integral@(occupancies*P_steady)))
    
    return mean_occupancy, std_occupancy
```

## Dependence of standard deviation on $J$

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

beta  = 1
htf   = -15
eb    = 11
L     = 4
sites = np.arange(L)
T     = 45000000

mean_occupancy = []
std_occupancy = []

J_list = np.linspace(0,15,151)
for J in J_list:
    #mu, std = calculate_mean_and_std_limit(htf, eb, J, L, sites, T)
    mu, std = calculate_mean_and_std(htf, eb, J, L, sites, T)
    mean_occupancy.append(mu)
    std_occupancy.append(std)
    
plt.plot(J_list, mean_occupancy, label='$\mu$ occupancy')
plt.plot(J_list, std_occupancy, label='$\sigma^2$ occupancy')
plt.xlabel('Clustering strength $J$')
plt.ylabel('Mean / Standard deviation of time average')
plt.ylim([-0.01,1.01])
plt.legend()
```

### Non-periodic grid

What if we do not have periodic BCs?

**Step 1: Finding eb to compensate for changes in J**

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

beta  = 1
htf   = -15
J     = 4
L     = 4
sites = np.arange(L)

T = 45000000

eb_list = np.linspace(0,15,151)
means = []
for eb in eb_list:
    mu, std = calculate_mean_and_std(htf, eb, J, L, sites, T, periodic=False)
    means.append(mu)
    
plt.figure()
plt.plot(eb_list, means)
plt.xlabel('Binding energy $e_b$')
plt.ylabel('Mean occupancy')
```

We see that there is only one value of $e_b$ for which the occupancy is 0.5, so we can use bisection to estimate this value of $e_b$.

```python
def bisection_eb(L, htf, eb_init, J, sites, T, tol = 0.02, periodic=False):
    # Bisection method
    while (eb_init[1]-eb_init[0]) > tol:

        eb = (eb_init[0]+eb_init[1])/2            
        occupancy, _ = calculate_mean_and_std(htf, eb, J, L, sites, T, periodic=periodic)

        if occupancy > 0.5:
            eb_init[1] = eb
        else:
            eb_init[0] = eb

    return (eb_init[0]+eb_init[1])/2 # Error = tol/2
```

```python
beta  = 1
htf   = -15
J     = 4
L     = 4
sites = np.arange(L)

T = 45000000

print('e_b =', bisection_eb(L, -15, [0,15], J, sites, T, tol = 0.02, periodic=False))
```

**Step 2: Analytically calculating the standard deviation for specific $h_{tf}$, $e_b$, $J$ and $T$.**

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

htf_to_c   = lambda htf : 1/(1+np.exp(-beta*htf))

beta  = 1
L     = 4
sites = np.arange(L)
T     = 45000000

mean_occupancy = []
std_occupancy = []

fig, ax = plt.subplots(1,2,figsize=(12,4))

#J_list = np.linspace(0,15,16)
J_list = [0, 3, 6, 9, 12, 15]
htf_list = np.linspace(-17,-13,41)
for j,J in enumerate(J_list):
    eb = bisection_eb(L, -15, [0,15], J, sites, T, tol = 0.02, periodic=False)
    
    mean_occupancy.append([])
    std_occupancy.append([])
    
    for htf in htf_list:
        mu, std = calculate_mean_and_std(htf, eb, J, L, sites, T, periodic=False)
        mean_occupancy[-1].append(mu)
        std_occupancy[-1].append(std)

    ax[0].plot(htf_list, mean_occupancy[-1], '--', color=f'C{j}')
    ax[0].plot(htf_list, std_occupancy[-1], color=f'C{j}', label=f'J={J}')
    ax[1].plot(htf_to_c(htf_list), mean_occupancy[-1], '--', color=f'C{j}')
    ax[1].plot(htf_to_c(htf_list), std_occupancy[-1], color=f'C{j}', label=f'J={J}')

Hill_eqn   = lambda c, c_05, n : 1/(1+(c_05/c)**n)
ax[1].plot(htf_to_c(htf_list), Hill_eqn(htf_to_c(htf_list), htf_to_c(-15), L), 'k:', label='Hill function')
    
ax[0].set_xlabel('Negative chemical potential $h_{tf}$')
ax[0].set_ylabel('Mean / Standard deviation of time average')
ax[0].set_ylim([-0.01,1.01])
ax[0].legend()

ax[1].set_xlabel('Concentration $c$ in molecules per grid cell')
ax[1].set_ylabel('Mean / Standard deviation of time average')
ax[1].set_ylim([-0.01,1.01])
ax[1].legend()

plt.figure()
for j,J in enumerate(J_list):
    plt.scatter(mean_occupancy[j], std_occupancy[j], color=f'C{j}', label=f'J={J}')
#plt.legend()
plt.xlabel('mean')
plt.ylabel('standard deviation')
plt.legend()
```

What is going on at $J=0$?

What if we do not change $e_b$, but instead look at the standard deviation at the shifted halfway point?

```python
def bisection_htf(L, htf_init, eb, J, sites, T, tol = 0.02, periodic=False):
    # Bisection method
    while (htf_init[1]-htf_init[0]) > tol:

        htf = (htf_init[0]+htf_init[1])/2            
        occupancy, _ = calculate_mean_and_std(htf, eb, J, L, sites, T, periodic=periodic)

        if occupancy > 0.5:
            htf_init[1] = htf
        else:
            htf_init[0] = htf

    return (htf_init[0]+htf_init[1])/2 # Error = tol/2
```

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

beta  = 1
L     = 4
htf   = -15
eb    = 11
sites = np.arange(L)
T     = 45000000

mean_occupancy = []
std_occupancy = []

J_list = np.linspace(0,15,16)
for j,J in enumerate(J_list):
    eb_05 = bisection_eb(L, htf, [0,-htf], J, sites, T, tol = 0.02, periodic=False)
    mu_eb, std_eb = calculate_mean_and_std(htf, eb_05, J, L, sites, T, periodic=False)
    htf_05 = bisection_htf(L, [-30,0], eb, J, sites, T, tol = 0.02, periodic=False)
    mu_htf, std_htf = calculate_mean_and_std(htf_05, eb, J, L, sites, T, periodic=False)
    mean_occupancy.append([mu_eb, mu_htf])
    std_occupancy.append([std_eb, std_htf])

plt.plot(J_list, std_occupancy)
plt.gca().set_prop_cycle(None)
plt.plot(J_list, mean_occupancy, '--')
plt.xlabel('Clustering strength $J$')
plt.ylabel('Mean / Standard deviation of time average')
plt.ylim([-0.01,1.01])
plt.title('Standard deviation at 0.5 occupancy')
plt.legend(['adapting $e_b$', 'adapting $h_{tf}$'])
```

So even if we take the standard deviation at the halfway point at different concentrations, it increases with J (provided that the range of $htf$ values used for the bisection method is lare enough). However, the way in which it increases does differ.

Adding non-binding sites seems to only have a significant effect for very large $J$, since only then those site might be filled. However, it allows for steeper occupancy curves and a more narrow standard deviation peak in the large $J$ limit. This suggests that if we have an infinite grid and infinite bath (sucht that the whole griid can fill up), the mean occupancy curve in steady state will indeed approach step function, and the standard deviation might approach a very narrow peak with max value 0.5 at the halfway point.
