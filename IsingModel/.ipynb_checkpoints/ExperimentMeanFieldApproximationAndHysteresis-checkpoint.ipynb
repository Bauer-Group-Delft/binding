# Mean field approximation

## Trajectories

```python
# Trajectories of the recursion relation for p.

beta = 1
htf = -3.5
N = 2
eb = 0.9
J = 2.6

p0 = np.exp(beta*htf)/(np.exp(beta*htf)+1)
p_map = lambda p : np.exp(beta*htf)/(np.exp(beta*htf)+np.exp(-beta*(eb+N*p*J)))
p_lower = np.exp(beta*htf)/(np.exp(beta*htf)+np.exp(-beta*eb))
p_upper = p_map(1)

ps = [np.array([p_lower, p_upper, 0.5])]
it = 100
for i in range(it):
    ps.append(p_map(ps[i]))

x = np.arange(it+1)
plt.figure()
plt.plot(x, ps)
plt.plot(x, np.full_like(np.array(ps)[:,0],p_upper), ':', label='upper bound')
plt.plot(x, np.full_like(np.array(ps)[:,0],p_lower), ':', label='lower bound')
plt.plot(x, np.full_like(np.array(ps)[:,0],p0), ':', label='fill fraction')
plt.legend()
```

Unstable fixed point at (htf=-3.5, p=0.5) (with eb=0.9 and J=2.6), because there $k_{on} = e^{-\beta h_{tf}}=k_{off}=e^{-\beta(e_b+nJ)}$, with $n=2p=1$.

Histeresis only exists when J>0, otherwise the upper bound and lower bound are equal and the prediction will be equal to those. When J>0, the lower and upper bound are different and the prediction will be very close to the lower bound for low htf and very close to the upper bound for high htf. 

The more clustering (higher J), the further the upper and lower bound are apart and the steeper the transition between them. For low values of J there is no histeresis yet. For values of J above a certain treshold (around 1.8 or 1.9), steep transition and hysteresis are observed. The higher the value of J past this threshold, the larger the region of histeresis.
The steepest response is achieved in the case of histeresis, but since the standard deviation should remain as low as possible to get a steep response, it would be best to have as small of a histeresis loop as possible, i.e. have the switching tresholds close together. This suggests an optimal J around 1.8-2.
However, a little bit of memory might be useful to prefent to rapidly switching on and off as a result of stochastic fluctuations in concentration (i.e. htf).

Since the model is stochastic we could expect random switching between the two equilibria. However, because we average the binding site occupancies from the model, we will not observe the jumps, but rather a gradual increase caused by the probability of obtaining the higher equilibrium increasing.

```python
from ipywidgets import interactive # for some neat interactions

def interactive_plot(htf=-1, eb=1, J=0):
    beta = 1
    N = 2
    p = np.linspace(0,1,100)
    
    fig, ax = plt.subplots(1,2, figsize=(10,4))
    
    ax[0].plot((1-p)*np.exp(beta*htf), p, label='$(1-p)\cdot k_{on}$')
    ax[0].plot(p*np.exp(-beta*(eb+p*N*J)), p, label='$p\cdot k_{off}(p)$')
    ax[0].legend()
    ax[0].set_xlim(left=0)
    ax[0].set_ylim([0,1])
    ax[0].set_xlabel('flows')
    ax[0].set_ylabel('$p$')
    
    plot_htf_prediction(beta, N, eb, J, ax=ax[1])
    ax[1].plot([htf,htf], [0,1], '--k')
    ax[1].set_xlim([-10,0])
    ax[1].set_ylim([0,1])

w = interactive(interactive_plot, htf=(-10.0,0.0,0.05), eb=(0.0,10.0,0.05), J=(0.0,10.0,0.05))
display(w)
```

The fixed points occur for the values of $p$ where the flows are equal.

# Comparison to model results

```python
from Experiments import average_occupancy_plot
from Predictions import plot_htf_prediction

beta = 1
N = 2
htf_list = np.linspace(-5,-3,11)
eb = 1
J = 3
T = 20000
runs = 20

L = 10
sites = np.arange(L) # All sites are binding sites.

fig, ax = plt.subplots()
average_occupancy_plot(runs, htf_list, (eb,T), 'htf', '$h_{tf}$', TF_Grid_1D, \
                       NSites=len(sites), L=L, J=J, sites=sites, predict=False, ax=ax)
plot_htf_prediction(beta, N, eb, J, ax=ax)
ax.set_xlim([min(htf_list),max(htf_list)])
ax.get_legend().remove()
```

The prediction seems much more steep than the simulation values. Why? Probably because the prediction is for all binding sites and the simulation is not. For all binding sites, it is closer, but the prediction is still steeper.

For J<1 the prediction seems to work well, but for values of J closer to the the point where hysteresis starts, the prediction becomes too steep. Could this have to do with too short run times? Larger J => larger dwell times => larger equilibration time.
Run time does not seem to make a difference.

Could it be that slight changes in local concentration even near the histeresis loop lead to locally a different effective htf, and thus a different occupancy? That would explain why the jump is less steep, cause jumping up might already occur at lower htf than predicted and jumping down at higher htf than predicted, leading to averaging and flattening of the steep response.
This indeed seems to be the case.

However the prediction being too steep also occurs for values of J between 1 and 2, most of which do not show hysteresis. If the response is steep enough the same explanation could still work, because then a small change in effective htf leads to a large variance.

A larger value of J leads to a more "all or nothing" response, however it will also take longer to switch, therefore a longer measurement time is required to get a similar standard deviation. There is likely an optimum value of J which depends on the execution time. If J is higher, the curve becomes more steep, so if T is large enough, a higher J would likely give higher precision. However, the chosen T limits the precision which can be obtained, therefore a less stepp curve might be better, since a high precision is reached faster there, even though the precision would never be as high as for larger J when run long enough.

# Comparison to Master equation solution

```python
from MarkovChain import solve_full_matrix_htf, plot_cumulative
from Predictions import plot_htf_prediction

## -------------------------------- Define parameters --------------------------------

beta = 1
N = 2
htf_list = np.linspace(-10,1,100)
eb   = 1
J    = 6

L   = 4
sites = np.arange(L)

T     = 1e10
steps = 1

v, occupancy = solve_full_matrix_htf(beta, htf_list, eb, J, L, sites, T, steps)
fig, ax = plot_cumulative(htf_list, v, '$h_{tf}$', [f'{i} TFs bound' for i in range(L+1)])
ax.plot(htf_list, occupancy, linestyle='--', c='k', label='occupancy') # expected fraction of binding sites occupied
ax.legend()
plot_htf_prediction(beta, N, eb, J, ax=ax)
```

For eb=0.9, J=6, L=6 (all binding sites) the gene expression response is very 'all or nothing' where wither all sites are filled or no sites are filled and the probability for an intermediate number of sites being filled is practically zero for all htfs. The switch from no sites to all sites being filled occurs gradually around htf=-7 by decreasing the probability of one while increasing the probability of the other.

For eb=0.9, J=2.6, L=6 (all binding sites) the gene expression response is more gradual, with a reasonable probability of an intermediate number of sites being filled around the switching point, which is not around htf=-3.5. Note that even though intermediate values are now more probable, it is still most likely that all or none of the sites are filled.

When J is high, the probability for intermediate values is low and the expected occupancy is almost exactly the probability fo the grid being fully occupied. When J is lower, the  probability of intermediate values is higher and the graph for the occupancy will be less steep than that for the probability of all sites being filled.

Note however that these results are for small grids, where the clustering can quickly cause the binding to spread over the whole grid. In that case the mean-field solution is accurate, but the periodic BCs are not realistic. However for larger grids, the mean field solution is no longer accurate, because it takes time for clusters to spread. (Is there a way to quantify this time?) We then expect the intermediate values to have higher probabilities.

It is likely not feasible to simulate larger grids with this model, however, maybe we will be able to find more information using the simplified matrix.
