```python
def determine_energy(s, binding_sites, htf, eb, J):
    E = 0
    for i in range(len(s)):
        E += -s[i]*(htf + int(i in binding_sites)*eb)
    E += -np.sum(s*np.concatenate([s[1:],[s[0]]]))*J
    return E

def calculate_probabilities(htf, eb, J):
    E = np.zeros(Nstates)
    for s_idx, s in enumerate(int_states):
        E[s_idx] = determine_energy(s, sites, htf, eb, J)
    p_steady = np.exp(-beta*E)
    return p_steady/np.sum(p_steady)

def calculate_state_occupancies(sites, normalize=True):
    # Calculate the mean occupancy of sites for each state

    sites = np.array(sites)
    state_occupancies = np.sum(int_states[:, sites], axis=1) # only works if sites is numpy array
    if normalize:
        state_occupancies = state_occupancies/len(sites)
    return state_occupancies

def compact_states(v, sites):
    # Combine states with the same number of TFs bound on the given sites.

    n = calculate_state_occupancies(sites, normalize=False)
    v_compact = np.zeros(len(sites)+1)
    for i in range(len(n)):
        v_compact[n[i]] += v[i]

    return v_compact
```

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

def calculate_mean_and_var(htf, eb, J, L, sites, T, mode = 'standard', periodic = True):
    
    # Set up matrix
    R = FullRates(mode, htf=htf, eb=eb, J=J)
    E = FullEnergies(mode, htf, eb, J)
    M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
    mat = M.create_Matrix(sparse=False)
    occupancies = M.states.calculate_state_occupancies(sites)
    P_steady = M.calculate_steady_state()

    # Mean occupancy
    mean_occupancy = occupancies@P_steady
    
    # Std occupancy
    eigenvalues, eigenvectors = np.linalg.eig(mat)   
    order = np.argsort(np.abs(eigenvalues))
    eigs = eigenvalues[order]
    P = eigenvectors[:,order]
    
    D_int = np.zeros((len(eigs),len(eigs)))
    D_int[1:,1:] = np.diag((-2/(eigs[1:]*T)**2)*(1-np.exp(eigs[1:]*T)+eigs[1:]*T))
    integral = P@D_int@np.linalg.inv(P)

    var_occupancy = occupancies@integral@(occupancies*P_steady)
    
    return mean_occupancy, var_occupancy.real # Note that occupancy can have a very small imaginary part
```

```python
N_bins = 100
p_occupation_given_htf = np.zeros((len(htf_list),N_bins))

for i,htf in enumerate(htf_list):
    # Plot histogram
    plt.figure()
    hist_values, bin_edges, _ = plt.hist(occupancies[i], N_bins, [0,1], density=True)
    plt.plot()
    plt.xlim([0,1])
    plt.title('$h_{tf}$ = '+f'{htf}')
    
    generator = steady_state_ICs(htf, eb, J)
    plt.plot(np.arange(len(sites)+1)/len(sites), compact_states(generator.P_steady,sites)/(1/(len(sites)+1)), 'o-', label='steady state distribution')
    plt.scatter([np.arange(len(sites)+1)@compact_states(generator.P_steady,sites)/len(sites)], [0], color='red', label='steady state mean')
    
    mu, sigma_squared = calculate_mean_and_var(htf, eb, J, L, sites, T)
    xs = np.linspace(0,1,1001)
    plt.plot(xs, sp.stats.norm.pdf(xs, mu, np.sqrt(sigma_squared)), 'k-')
    
    # Store distribution
    p_occupation_given_htf[i,:] = hist_values*(1/N_bins) # Discretized distribution

print(np.sum(p_occupation_given_htf, axis=1))
#assert(np.all(np.sum(p_occupation_given_htf, axis=1)==1))
```

# Mutual information from steady state distribution ($T=0$)

```python
def calculate_boltzmann_factors(S, E, sites, beta=1):
    energies = np.zeros(S.Nstates)
    for s_idx, s in enumerate(S.int_states):
        energies[s_idx] = E.energy_function(s, sites, periodic=False)
    return np.exp(-beta*energies)

def calculate_steady_state_mutual_information(htf_list, eb, J, L, sites, beta=1, plot=False, mode=1):
    P_occupancy_given_htf = np.zeros((len(htf_list),len(sites)+1))
    P_occupancy_and_htf = np.zeros((len(htf_list),len(sites)+1))
    P_occupancy = np.zeros(len(sites)+1)

    for i,htf in enumerate(htf_list):
        S = FullStates(L)
        E_tot = FullEnergies('standard', htf, eb, J)

        # P(occupancy|htf)
        boltzmann_factors = calculate_boltzmann_factors(S, E_tot, sites, beta)
        partition_function = np.sum(boltzmann_factors)
        boltzmann_factors_per_occupancy = S.compact_states(np.array([boltzmann_factors]), sites)[0]
        P_occupancy_given_htf[i,:] = 1/partition_function*boltzmann_factors_per_occupancy

        # P(occupancy,htf)
        P_htf = 1/len(htf_list)
        P_occupancy_and_htf[i,:] = P_occupancy_given_htf[i,:]*P_htf

        # P(occupancy)
        P_occupancy += P_occupancy_and_htf[i,:]
        
    if plot:
        plt.figure()
        plt.imshow(P_occupancy_given_htf, aspect='auto', interpolation='none')
        plt.title(f'$J={J}$, $e_b={eb}$')

    I1 = np.sum(P_occupancy_and_htf*np.log2(P_occupancy_given_htf/P_occupancy))
    
    if mode == 1:
        return I1
    
    P_occupancy_given_htf_small = np.sum(P_occupancy_given_htf[np.where(htf_list<-15)[0],:], axis=0)/len(np.where(htf_list<-15)[0])
    P_occupancy_given_htf_large = np.sum(P_occupancy_given_htf[np.where(htf_list>=-15)[0],:], axis=0)/len(np.where(htf_list>=-15)[0])
    # print(np.sum(P_occupancy_given_htf_small))
    # print(np.sum(P_occupancy_given_htf_large))
    P_occupancy_and_htf_small = np.sum(P_occupancy_and_htf[np.where(htf_list<-15)[0],:], axis=0)
    P_occupancy_and_htf_large = np.sum(P_occupancy_and_htf[np.where(htf_list>=-15)[0],:], axis=0)
    # print(np.sum(P_occupancy_and_htf_small+P_occupancy_and_htf_large))
    
    I2 = np.sum(P_occupancy_and_htf_small*np.log2(P_occupancy_given_htf_small/P_occupancy) + P_occupancy_and_htf_large*np.log2(P_occupancy_given_htf_large/P_occupancy))
    
    if mode == 2:
        return I2
    elif mode==0:
        return I1, I2
        
```

If we do not try to compensate for the shift, we obtain a peak around the value of J for which the occupancy curve is centered in the relevant conentration range. This indicates that shifting the occupancy plot decreases the mutual information, because the part that contains the most information shifts out of the relevant concentration range (compare to below).

As $J$ increases, the effect of htf on the occupancy becomes stronger. At low $J$, the occupancy is almost always very close to zero. The higher $J$ gets, the more likely the occupancy is to be non-zero, expecially for higher htf values. Therefore, the low htf values are still likely to give an occupancy of zero, while high values can now be distinguished, since they lead to an occupancy of one. The higher $J$ becomes, the more htf values will have a most likely occupancy of 1. This will increase the certainty with which the htf values can be distinquished, untill there are equally many htf values with most likely occupancy 0 and 1. Since increasing $J$ any further would lead to many htf values having occupancy 1 and those would therefore become less destinguishable. The mutual information thus decreases again when $J$ is increased further. 

What if we do compensate?

**Using $e_b$ to compensate for shift:**

```python
from MarkovChainClass import FullStates, FullEnergies

def calculate_mean_occupancy(htf, eb, J, L, sites, beta=1):
    S = FullStates(L)
    E_tot = FullEnergies('standard', htf, eb, J)

    # P(occupancy|htf)
    boltzmann_factors = calculate_boltzmann_factors(S, E_tot, sites, beta)
    partition_function = np.sum(boltzmann_factors)
    boltzmann_factors_per_occupancy = S.compact_states(np.array([boltzmann_factors]), sites)[0]
    P_occupancy_given_htf = 1/partition_function*boltzmann_factors_per_occupancy
    
    return P_occupancy_given_htf@np.arange(len(sites)+1)/len(sites)

def bisection_eb(L, htf, eb_init, J, sites, tol = 0.02):
    # Bisection method
    while (eb_init[1]-eb_init[0]) > tol:

        eb = (eb_init[0]+eb_init[1])/2            
        occupancy = calculate_mean_occupancy(htf, eb, J, L, sites)

        if occupancy > 0.5:
            eb_init[1] = eb
        else:
            eb_init[0] = eb

    return (eb_init[0]+eb_init[1])/2 # Error = tol/2
```

```python
L     = 6
sites = np.arange(L)

htf_list = np.linspace(-13,-17,401)
J_list = np.linspace(0,15,16)
I = np.zeros(len(J_list))
I2 = np.zeros(len(J_list))
for j, J in enumerate(J_list):
    eb = bisection_eb(L, -15, [0,15], J, sites, tol = 0.002)
    print(calculate_mean_occupancy(-15, eb, J, L, sites))
    print(J, eb)
    I[j], I2[j] = calculate_steady_state_mutual_information(htf_list, eb, J, L, sites, plot=False, mode=0)
    clear_output(wait=True)

fig, ax = plt.subplots(1,2, figsize=(8,3))
fig.tight_layout(pad=5.0)
ax[0].plot(J_list, I)
ax[0].set_xlabel('Clustering strength $J$')
ax[0].set_ylabel('Mutual information $I$')
ax[0].set_title('All htf values')

ax[1].plot(J_list, I2, ':')
ax[1].set_xlabel('Clustering strength $J$')
ax[1].set_ylabel('Mutual information $I$')
ax[1].set_title('htf < or >= -15')
```

If we do compensate for the shift, the mutual infromation initially increases with $J$ after which it slowly decreases again and seems to go to a certain limit. A possible explanation for the initial increase is that the overlap in the conditional probability distributions of the occupancy decreases, leading to less possible occupancy values for each htf. However, when most of the htf values are assigned an occupancy of 0 or 1, it will again be harder to distinguish them, leading to a slow decrease in standard deviation. This corresponds to the occupancy plot becoming more steep. Initially an increase in steepness will make the htfs around the halfway point more distinguishable, leading to an increase in mutual information, while if the steepness increases too much the range of htf values which can be well destinguished becomes very small, the standard deviation also increases, and the mutual information decreases again. (This suggests that for small J the increase in steepness outcompetes the increase in standard deviation caused by increasing $J$.)

If we only consider whether the nucleus is in the front or the back of the embryo, the mutual information increases with $J$, untill it approaches a certain limit. In this case it does not matter that different very high htf values (or very low htf values) can not be clearly sitinguished from each other, since they are all in the front (or back) of the embryo anyway. So the steeper the curve around the halfway point, the better (as long as the standard deviation does not become too large).

If we increase $J$ too much, we can not decrease $e_b$ further to compensate without making it negative. 

The mutual infromation is bounded from above by the entropy of the occupancy probability distribution, which in this case is low, because there are only $L+1$ possible values (discrete distribution).

*Adding a non-binding site padding:*

```python
N = 6
padding = 2
L = N+2*padding
sites = np.arange(N)+padding

htf_list = np.linspace(-13,-17,401)
J_list = np.linspace(0,15,16)
I = np.zeros(len(J_list))
I2 = np.zeros(len(J_list))
for j, J in enumerate(J_list):
    eb = bisection_eb(L, -15, [0,15], J, sites, tol = 0.002)
    print(calculate_mean_occupancy(-15, eb, J, L, sites))
    print(J, eb)
    I[j], I2[j] = calculate_steady_state_mutual_information(htf_list, eb, J, L, sites, plot=False, mode=0)
    clear_output(wait=True)

fig, ax = plt.subplots(1,2, figsize=(8,3))
fig.tight_layout(pad=5.0)
ax[0].plot(J_list, I)
ax[0].set_xlabel('Clustering strength $J$')
ax[0].set_ylabel('Mutual information $I$')
ax[0].set_title('All htf values')

ax[1].plot(J_list, I2, ':')
ax[1].set_xlabel('Clustering strength $J$')
ax[1].set_ylabel('Mutual information $I$')
ax[1].set_title('htf < or >= -15')
```

Note how the non-binding sites only significantly affect the graph for very high $J$, since only then the binding sites will be filled and assuming the binding sites to be empty will not be a good approximation. However, the extra stabilization provided by the presence of TFs on the non-binding sites (at very high J) seems to increase the mutual information even when all htf values are considered. A possible explanation is that for only binding sites, the occupancy curve at some point aproaches a hill equation with as hill coefficient the number of bidnign sites, so the maximum steepness of the curve is limited, after which increasing $J$ no longer increases the steepness but only increases the standard deviation, therefore decreasing the mutual information (in the case of all htf values). However, if the steepness can increase further, once $J$ becomes very high, the mutual information can go up again at those values of $J$. 

### Comparing Mutual information to steepness of occupancy curve

```python
from MarkovChain import solve_steady_state_htf

## -------------------------------- Define parameters --------------------------------

beta  = 1
L     = 6
sites = np.arange(L)

htf_to_c = lambda htf : 1/(1+np.exp(-beta*htf))
#c_to_htf = lambda c : 1/beta*(np.log(c)-np.log(1-c))

htf_05 = -15
c_05 = htf_to_c(htf_05)
htf_list = np.linspace(-17,-13,401)
conc_range = htf_to_c(htf_list)

Hill_eqn = lambda c, n : 1/(1+(c_05/c)**n)         # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)

J_list = np.linspace(0,15,16)
n_opt = np.zeros(len(J_list))
for j,J in enumerate(J_list):
    eb = bisection_eb(L, htf_05, [0,-htf_05], J, sites, tol = 0.002)
    
    ## -------------------------------- Calculate solution --------------------------------
    
    v_compact, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites, periodic=False)

    ## -------------------------------- Fit hill curve --------------------------------

    n_opt[j], _ = sp.optimize.curve_fit(Hill_eqn, conc_range, occupancy, bounds = ([0,100]))
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

plt.figure()
plt.plot(J_list,n_opt)
plt.xlabel('J')
plt.ylabel('Optimal Hill coefficient, n')
plt.xlim([J_list[0],J_list[-1]])
plt.ylim(bottom=0)
```

We see that without the non-binding site padding, the optimal hill coefficient of teh fit increases, but seems to approach a limit. While in the presense of the non-binding site padding, the curve initially also seems to approach this limit, but for very high $J$ increases faster again. The shapes of these curves resemble the shape of the curves of the mutual information, especially the one where we only consider htf < or >= -15. This indicates that a large part of the observed behaviour in the mutual information graphs is related to the steepness of the occupancy curve. But likely the increasing standard deviation makes the mutual information curves steeper for low J and less steep for intermediate/high J.

# Mutual information from Gaussian around steady state mean ($T\rightarrow\infty$)

```python
import math as m
from scipy import special

def calculate_gaussian_mutual_information(htf_list, eb, J, L, sites, T, N=100, beta=1, plot=False, mode=1):
    P_occupancy_given_htf = np.zeros((len(htf_list),N))
    P_occupancy_and_htf = np.zeros((len(htf_list),N))
    P_occupancy = np.zeros(N)
    xs = np.linspace(0,1,N+1)

    for i,htf in enumerate(htf_list):
        # P(occupancy|htf)
        mu, var = calculate_mean_and_var(htf, eb, J, L, sites, T, periodic=False)
        # P_occupancy_given_htf[i,:] = sp.stats.norm.pdf(xs, mu, var)*N 
        # divide by 1/N (becomes *N) to go for continuous to discrete distribution.
        P_occupancy_given_htf[i,:] = special.erf((xs[1:]-mu)/np.sqrt(2*var))-special.erf((xs[:-1]-mu)/np.sqrt(2*var)) 
        
        # P(occupancy,htf)
        P_htf = 1/len(htf_list)
        P_occupancy_and_htf[i,:] = P_occupancy_given_htf[i,:]*P_htf

        # P(occupancy)
        P_occupancy += P_occupancy_and_htf[i,:]
        
    if plot:
        plt.figure()
        plt.imshow(P_occupancy_given_htf, aspect='auto', interpolation='none')
        plt.title(f'$J={J}$, $e_b={eb}$')

    nonzero_terms = np.where(P_occupancy_and_htf!=0)
    I1 = np.sum(P_occupancy_and_htf[nonzero_terms]*np.log2(P_occupancy_given_htf[nonzero_terms]/P_occupancy[nonzero_terms[1]]))
    # If P_occupancy_and_htf is nonzero, so will P_occupancy_given_htf and P_occupancy, 
    # so we whould only take the logarithms of those terms, since all others are zero.
    
    if mode == 1:
        return I1
    
    P_occupancy_given_htf_small = np.sum(P_occupancy_given_htf[np.where(htf_list<-15)[0],:], axis=0)/len(np.where(htf_list<-15)[0])
    P_occupancy_given_htf_large = np.sum(P_occupancy_given_htf[np.where(htf_list>=-15)[0],:], axis=0)/len(np.where(htf_list>=-15)[0])
    # print(np.sum(P_occupancy_given_htf_small))
    # print(np.sum(P_occupancy_given_htf_large))
    P_occupancy_and_htf_small = np.sum(P_occupancy_and_htf[np.where(htf_list<-15)[0],:], axis=0)
    P_occupancy_and_htf_large = np.sum(P_occupancy_and_htf[np.where(htf_list>=-15)[0],:], axis=0)
    # print(np.sum(P_occupancy_and_htf_small+P_occupancy_and_htf_large))
    
    I2 = np.sum(P_occupancy_and_htf_small*np.log2(P_occupancy_given_htf_small/P_occupancy) + P_occupancy_and_htf_large*np.log2(P_occupancy_given_htf_large/P_occupancy))
    
    if mode == 2:
        return I2
    elif mode==0:
        return I1, I2
```

```python
L     = 6
sites = np.arange(L)
T     = 100*45000000

htf_list = np.linspace(-13,-17,401)
J_list = np.linspace(0,15,16)
I = np.zeros(len(J_list))
I2 = np.zeros(len(J_list))
for j, J in enumerate(J_list):
    eb = bisection_eb(L, -15, [0,15], J, sites, tol = 0.002)
    print(calculate_mean_occupancy(-15, eb, J, L, sites))
    print(J, eb)
    I[j], I2[j] = calculate_gaussian_mutual_information(htf_list, eb, J, L, sites, T, N=100, plot=True, mode=0)
    clear_output(wait=True)
    
fig, ax = plt.subplots(1,2, figsize=(8,3))
fig.tight_layout(pad=5.0)
ax[0].plot(J_list, I)
ax[0].set_xlabel('Clustering strength $J$')
ax[0].set_ylabel('Mutual information $I$')
ax[0].set_title('All htf values')

ax[1].plot(J_list, I2, ':')
ax[1].set_xlabel('Clustering strength $J$')
ax[1].set_ylabel('Mutual information $I$')
ax[1].set_title('htf < or >= -15')
```

Now the mutual information seems to decrease with $J$. I expect that if the measurement time is long, clustering is no longer beneficial, because there is enough time to measure the concentreation accurately and clustering only increases the standard deviation, making the measurements less accurate.

# Hill coefficient

```python
from MarkovChain import solve_steady_state_htf

## -------------------------------- Define parameters --------------------------------

beta  = 1
eb    = 6
L     = 6
sites = np.arange(L)

htf_to_c = lambda htf : 1/(1+np.exp(-beta*htf))
c_to_htf = lambda c : 1/beta*(np.log(c)-np.log(1-c))

J_list = np.linspace(0,10,21)
n_opt = np.zeros(len(J_list))

for j,J in enumerate(J_list):
    htf_05 = -(eb+J)
    htf_list = np.linspace(htf_05-6,htf_05+6,101)
    conc_range = htf_to_c(htf_list)
    
    ## -------------------------------- Calculate solution --------------------------------
    
    v_compact, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites)

    ## -------------------------------- Fit hill curve --------------------------------

    c_05 = htf_to_c(htf_05)
    Hill_eqn = lambda c, n : 1/(1+(c_05/c)**n)         # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)

    n_opt[j], _ = sp.optimize.curve_fit(Hill_eqn, conc_range, occupancy, bounds = ([0,100]))
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

plt.figure()
plt.plot(J_list,n_opt)
plt.xlabel('J')
plt.ylabel('Optimal Hill coefficient, n')
plt.xlim([J_list[0],J_list[-1]])
plt.ylim(bottom=0)
```
