# Cluster dynamics

Research question: Will multiple small drowplets or one large droplet form?

## Without depletion

```python
from Experiments import plot_kymograph_1D

htf = -8
eb_list  = [0, 2, 6, 10]
J   = 6
L   = 100
sites = np.arange(L)
T   = 200000

for eb in eb_list:
    plot_kymograph_1D(htf, eb, J, L, sites, T, steps = 1000)
```

```python
from Experiments import plot_kymograph_1D

htf = -8
eb  = 2
J_list   = [0, 2, 6, 10]
L   = 100
sites = np.arange(L)
T   = 200000

for J in J_list:
    plot_kymograph_1D(htf, eb, J, L, sites, T, steps = 1000)
```

For $J=10$ we see that initially multiple small droplets form. These then grow and merge together to cover the whole field. The fact that these droplets can keep growing like that is because there is an infinite bath of TFs. When implementing a bath with a fixed number of TFs, depletion of TFs will limit the growth of the droplets. There are then two scenarios, either multiple small droplets will exist simultaneously, or the smaller droplets disappear as the larger droplet grows and in the end only one large droplet will remain.

The growth of the larger droplet can have different causes:
- The system has a lower energy when there is one large droplet, compared to multiple small droplets, since more TFs will be inside the droplet and have more neighbors to interact with. This effect shouls also be observed in 1D.
- The Large droplet has a larger surface and can thus grow faster. This effect can not be observed in 1D since the boundary of each cluster in 1D is equal in size.
- However, large clusters might also shrink faster due to larger volume/surface. Only the volume part of this effect can be observed in 1D. However, in reality molecules disappear from droplets at the surface, so this volume effect might not be realistic.

## With depletion

```python
from Experiments import plot_kymograph_1D

htf = -8
eb  = 2
J_list   = [0, 6, 8, 10]
L   = 100
sites = np.arange(L)
T   = 10000000 #2000000

for J in J_list:
    plot_kymograph_1D(htf, eb, J, L, sites, T, steps = 1000, bath=True, R=0.1)
```

Lower $J$ means clusters fall apart faster and thus makes the clusters more dynamic.

The larger clusters seem quite stable, perhaps we can speed things up by tweaking the parameters.

```python
from Experiments import plot_kymograph_1D

htf = -8
eb_list = [0, 2, 6, 10]
J   = 8
L   = 100
sites = np.arange(L)
T   = 10000000 #2000000

for eb in eb_list:
    plot_kymograph_1D(htf, eb, J, L, sites, T, steps = 1000, bath=True, R=0.1)
```

Higher $e_b$ seems to lead to a larger number of smaller clusters, since nucleation of new clusters is more likely when TFs remain on the grid longer.
