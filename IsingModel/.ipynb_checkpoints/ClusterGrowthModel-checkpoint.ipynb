```python
import numpy as np
import matplotlib.pyplot as plt
```

# Probability of cluster size for a fixed number of molecules

```python
beta = 1
Nmolecules = 1000
Nsites = 1000**3
print(f'If NJ>{-(1/beta)*np.log(Nmolecules/Nsites)} there will be an optimal cluster size larger than 1.')
NJ = 13.9
k_off = np.exp(-beta*NJ)
#ratios = [((Nmolecules-i)/Nsites)/k_off for i in range(1,Nmolecules)]
ratios = [((Nmolecules-i)/(Nsites-i))/k_off for i in range(1,Nmolecules)]
#print(ratios)
P_n = np.concatenate(([1],np.cumprod(ratios)))
#print(P_n)
P_n = P_n/np.sum(P_n)
#print(P_n)

plt.plot(np.arange(1,Nmolecules+1), P_n)

#s = Nmolecules - Nsites*k_off
s = (Nmolecules - Nsites*k_off)/(1-k_off)
print('initial concentration:', Nmolecules/Nsites)
print('solubility:', k_off) # = (Nmolecules-s)/(Nsites-s)
print('optimal cluster size:', s)

plt.plot([s,s],[0,0.015], ':k')
plt.xlabel('cluster size')
plt.ylabel('probability in steady state')
```

In this case there is one cluster that grows or shrinks, but can we make multiple clusters compete?

The on-rate times the cocnentration is the inward flux. This flux is proportional to the concentration and inversely proportional to the radius. Should the outward flux also depend on the radius?

The inward flux likely depends on the radius, since if the radius is small, the flow from a large area is compressed near the surface to become a larger flow. A similar thing should be happening for the outward flux. Can we chech this by doing calculations for a spherical source?

```python
beta = 1
htf_to_c   = lambda htf : 1/(1+np.exp(-beta*htf))

htf_list = np.linspace(-17,-13)
c = htf_to_c(htf_list)
Nsites = 1000**3
Nmolecules = c*Nsites
NJ = 15
k_off = np.exp(-beta*NJ)
s = (Nmolecules - Nsites*k_off)/(1-k_off)

plt.plot(htf_list, [max(0,si) for si in s])
plt.xlabel('$h_{tf}$')
plt.ylabel('Cluster size')
plt.title('Most likely cluster size at steady state')
```

# Diffusion limited cluster growth

```python
beta = 1
Nmolecules = 1000
Nsites = 1000**3
D = 0.3 / 0.0004**2 # site^2/s
NJ = 2.5
k_off = np.exp(-beta*NJ)
ratios = [D*((Nmolecules-i)/(Nsites-i))/(i**(1/3)*k_off) for i in range(1,Nmolecules)]
#print(ratios)
P_n = np.concatenate(([1],np.cumprod(ratios)))
#print(P_n)
P_n = P_n/np.sum(P_n)
#print(P_n)

plt.plot(np.arange(1,Nmolecules+1), P_n)

plt.xlabel('cluster size')
plt.ylabel('probability in steady state')
```
