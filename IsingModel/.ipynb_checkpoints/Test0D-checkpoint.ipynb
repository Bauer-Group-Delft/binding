# Experiment 2: Dwell time histogram

```python
from Tests import simulate, exponential_distribution_histogram
from Predictions import predict_times

def dwell_time_histogram(datafile, T_app, T_disapp):
    t,grids = datafile.get_grids()
    dts = t[1:]-t[:-1]
    dwell_times = dts[np.where(grids[:-1])]
    binding_times = dts[np.where(np.logical_not(grids[:-1]))]
    
    fig, axs = plt.subplots(1, 2)
    fig.tight_layout(pad=3)
    
    exponential_distribution_histogram(axs[0], dwell_times, T_disapp, 'Dwell times', 'Prediction: $P=\\frac{1}{\\tau_{d}}e^{-t/\\tau_{d}}$')
    exponential_distribution_histogram(axs[1], binding_times, T_app, 'Times till binding', 'Prediction: $P=\\frac{1}{\\tau_{a}}e^{-t/\\tau_{a}}$')

htf   = -1
eb    = 1
timesteps = 20000

datafile = simulate(TF_Grid_0D, np.inf, htf, eb, timesteps)
T_app, T_disapp = predict_times(htf, eb)
clear_output(wait=True)

dwell_time_histogram(datafile, T_app, T_disapp)
print(f'tau_d = {T_disapp}, tau_a = {T_app}.')
```

## Experiment 3.2: Average occupancy against $h_{tf}$

```python
from Experiments import average_occupancy_plot

htf_list = np.arange(-10,1,2)
eb = 10
T = 20000
runs = 100

average_occupancy_plot(runs, htf_list, (eb,T), 'htf', '$h_{tf}$', TF_Grid_0D)
clear_output(wait=True)
```

If the runtime is too short or the number of runs to small, the obtained mean and standard deviation are not accurate. 

Note: The initial fill fraction is calculated based on k_off = 1, i.e. eb=0. If the eb is nonzero, the mean occupation will be higher than the fill fraction, however, it will take time for the system to equilibrate and reach this mean occupancy. This time depends on the on and off rates. If the on and off rates are low, the equilibration time will be long and results obtained for short runtimes will be inaccurate. This effect is most noticable when the on-rate is low (low htf), leading to a very small fill fraction, but the off-rate is also very low (high eb), leading to a large expected occupancy.
