```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

# Matrix exponential integrals

```python
def matrix_exponential_integral(mat, T, mode=1):
    if mode == 0: # Invertible matrix
        I = np.identity(len(mat))
        integral = np.linalg.inv(mat)@(sp.linalg.expm(mat*T)-I)
    
    elif mode == 1: # Matrix with one eigenvalue equal to 0
        eigenvalues, eigenvectors = np.linalg.eig(mat)
        
        order = np.argsort(np.abs(eigenvalues))
        # print('eigenvalues:', eigenvalues[order])
        D = np.diag(eigenvalues[order])
        P = eigenvectors[:,order]
        
        ## Test: accuracy of diagonalization, PDP^-1 should be equal to the original matrix.
        # print(mat)
        # print(eigenvectors@np.diag(eigenvalues)@np.linalg.inv(eigenvectors))
        # print(P@D@np.linalg.inv(P))
        ## --------------------------------------------------------------------------------
        
        D_int = np.zeros_like(D)
        D_int[0,0] = T
        D_int[1:,1:] = matrix_exponential_integral(D[1:,1:], T, mode=0)

        integral = P@D_int@np.linalg.inv(P)
    
    return integral
```

```python
def matrix_exponential_integral_limit(mat, T_inf=0):
    eigenvalues, eigenvectors = np.linalg.eig(mat)
    order = np.argsort(np.abs(eigenvalues))
    P = eigenvectors[:,order]

    D_int = np.zeros((len(eigenvalues),len(eigenvalues)))
    D_int[0,0] = T_inf
    D_int[1:,1:] = np.diag(-1/(eigenvalues[order])[1:])

    integral = P@D_int@np.linalg.inv(P)
    
    return integral
```

*Setup matrix:*

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

def setup_matrix(beta = 1, htf = -15, eb = 11, J = 4, L = 4, sites = None, mode = 'standard', periodic = True):
    if sites == None:
        sites = np.arange(L)
        
    R = FullRates(mode, htf=htf, eb=eb, J=J)
    E = FullEnergies(mode, htf, eb, J)
    M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
    mat = M.create_Matrix(sparse=False)
    occupancies = M.states.calculate_state_occupancies(sites)
    grid_cell_states = M.int_states.T
    P_steady = M.calculate_steady_state()
    
    return mat, occupancies, grid_cell_states, P_steady
```

# Mean and standard deviation of occupancy

*Mean and standard deviation of occupancy in steady state:*

```python
mat, occupancies, grid_cell_states, P_steady = setup_matrix()

mu = occupancies@P_steady
mu_j = grid_cell_states@P_steady
print('In steady state:')
print('- Mean occupancy:', mu, '\n- Mean grid cell state', mu_j)

sigma_squared = (occupancies-mu)**2@P_steady
sigma_squared_j = (np.array([grid_cell_states[i]-mu_j[i] for i in range(L)]))**2@P_steady
print('- Variance of occupancy:', sigma_squared, '\n- Variance of grid cell state:', sigma_squared_j)
```

# Time average of occupancy

*Time average of occupancy in steady state:*

```python
T = 45000000
mat, occupancies, grid_cell_states, P_steady = setup_matrix()
integral = matrix_exponential_integral(mat, T)

mean_occupancy = (occupancies@(integral@P_steady))/T
mean_occupancy_per_site = (grid_cell_states@(integral@P_steady))/T
print('In steady state:')
print('- Average of mean occupancy over time T:', mean_occupancy)
print('- Average of grid cells states over time T:', mean_occupancy_per_site)

# Note: In steady state the time integral should give the same result as the mean occupancy at any time.
```

*Time average of occupancy starting from v0:*

```python
T = 45000000
mat, occupancies, grid_cell_states, P_steady = setup_matrix()
integral = matrix_exponential_integral(mat, T)
v0 = np.zeros(len(mat))
v0[0]=1

integral = matrix_exponential_integral(mat, T)
mean_occupancy = (occupancies@(integral@v0))/T
mean_occupancy_per_site = (grid_cell_states@(integral@v0))/T
print('Starting from v0:')
print('- Average of mean occupancy over time T:', mean_occupancy)
print('- Average of grid cells states over time T:', mean_occupancy_per_site, '\n')
```

This seems to work for small matrices for which the calculated eigenvalues are real, but for $L>=5$ and certain parameter values (some of) the eigenvalues become imaginary.

We see that the time average of the occupancy is the same as the time average of the grid cells state of each grid, which is as expected, since all grid cells are identical.

# Variance in time average of occupancy

```python
L = 4
T = 45000000
mat, occupancies, grid_cell_states, P_steady = setup_matrix(L=L)
integral = matrix_exponential_integral_limit(mat, 0)

correlation_integral = np.zeros(L)
for j in range(L):
    correlation_integral[j] = grid_cell_states[j]@integral@(grid_cell_states[j]*P_steady)
variance = 2/T*correlation_integral
mu = grid_cell_states@P_steady
sigma_squared = (np.array([grid_cell_states[i]-mu[i] for i in range(L)]))**2@P_steady
tau = 1/sigma_squared*correlation_integral
print('Variance of average grid cell state over time T:', variance)
print(f'We need T={T} >> tau={tau}.\n')

correlation_integral_occupancy = occupancies@integral@(occupancies*P_steady)
variance_occupancy = 2/T*correlation_integral_occupancy
mu_occupancy = occupancies@P_steady
sigma_squared_occupancy = (occupancies-mu_occupancy)**2@P_steady
tau_occupancy = 1/sigma_squared_occupancy*correlation_integral_occupancy
print('Variance of average occupancy over time T:', variance_occupancy)
print(f'We need T={T} >> tau={tau_occupancy}.\n')

plt.plot(np.arange(L), variance, label=f'grid cell state')
plt.plot([0,L-1], [variance_occupancy]*2, ':', label='occupancy')
plt.xlabel('grid cell')
plt.ylabel('Variance in average occupancy')
plt.legend()
```

Note that using the occupancy to calculate the variance gives a different result than using the grid cells state! The grid cell states have a larger variance, since those states can only take the values 0 and 1, while the occupancy can also take values in between (i.e. closer to the mean). 

Note how the occupancy has a longer correlation time than the grid cells, since even if a grid cell is equilibrated, the states of the other grid cells might still be influenced by the past state of that grid cell.

We can compare this to the model, provided that we start the model in steady state.

The correlation time for a 2-state model, with rates corresponding to the entire cluster appearing and disappearing at once ($k_{on} = \exp(\beta Nh_{tf})$ and $k_{off} = \exp(-\beta N(eb+J))$), is much larger, becuase it is not possible to gradually build up or break down clusters to reach the other state.

```python
def calculate_means_stds(htf, eb, J, L, sites, T, mode = 'standard', periodic = True):
    
    R = FullRates(mode, htf=htf, eb=eb, J=J)
    E = FullEnergies(mode, htf, eb, J)
    M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
    mat = M.create_Matrix(sparse=False)
    occupancies = M.states.calculate_state_occupancies(sites)
    grid_cell_states = M.int_states.T
    P_steady = M.calculate_steady_state()

    mean_grid_cell_state = np.mean(grid_cell_states@P_steady)
    mean_occupancy = occupancies@P_steady
    
    integral = matrix_exponential_integral_limit(mat, 0)
    correlation_integrals = np.zeros(L)
    for i in range(L):
        correlation_integrals[i] = grid_cell_states[i]@integral@(grid_cell_states[i]*P_steady)
    std_grid_cell_state = np.mean(np.sqrt(2/T*correlation_integrals)) # mean of standard deviation of different grid cells
    std_occupancy = np.sqrt(2/T*(occupancies@integral@(occupancies*P_steady)))
    
    return mean_grid_cell_state, mean_occupancy, std_grid_cell_state, std_occupancy
```

# Equilibration time 

```python
beta  = 1
htf   = -15
#eb    = 11
L     = 4
T     = 45000000
sites = np.arange(L)

mode = 'standard'
periodic = True

J_list = np.arange(0,-htf+1)
tau_occupancy = []

for J in J_list:
    eb = -htf-J
    
    R = FullRates(mode, htf=htf, eb=eb, J=J)
    E = FullEnergies(mode, htf, eb, J)
    M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
    mat = M.create_Matrix(sparse=False)
    occupancies = M.states.calculate_state_occupancies(sites)
    grid_cell_states = M.int_states.T
    P_steady = M.calculate_steady_state()

    integral = matrix_exponential_integral_limit(mat, 0)
    correlation_integral_occupancy = occupancies@integral@(occupancies*P_steady)
    mu_occupancy = occupancies@P_steady
    sigma_squared_occupancy = (occupancies-mu_occupancy)**2@P_steady
    tau_occupancy.append(1/sigma_squared_occupancy*correlation_integral_occupancy)

plt.plot(J_list, tau_occupancy)
plt.xlabel('Clustering strength $J$')
plt.ylabel('Correlation time of occupancy')
```

If $e_b$ is adapted such that for the value of $J$, the occupancy is 0.5, then the correlation time increases with $J$.

So if we only consider the standard eviation (at the 0.5 occupancy point) we would like $J$ to be as low as possible. But that means the occupancy curve is less steep. So the optimal accuracy is probably a trade-off between low variance and high steepness. To know more about how this trade-off works exactly, we should calculate the mutual information.

# Correlation functions of occupany and grid cell state

```python
from MarkovChainClass import plot_analytical_correlation_function

beta  = 1
htf   = -15
eb    = 11
J     = 4
L     = 4
sites = np.arange(L)

T = 10*45000000
steps = 1000

mode = 'standard'
periodic = True

R = FullRates(mode, htf=htf, eb=eb, J=J)
E = FullEnergies(mode, htf, eb, J)
M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
mat = M.create_Matrix(sparse=False)
occupancies = M.states.calculate_state_occupancies(sites)
P_steady = M.calculate_steady_state()

integral = matrix_exponential_integral_limit(mat, 0)
correlation_integral_occupancy = (occupancies@integral@(occupancies*P_steady))
mu_occupancy = occupancies@P_steady
sigma_squared_occupancy = (occupancies-mu_occupancy)**2@P_steady
tau_occupancy = 1/sigma_squared_occupancy*correlation_integral_occupancy

fig, ax = plt.subplots()
plot_analytical_correlation_function(L, htf, eb, J, sites, T, steps, mode='occupancy', ax=ax)
plot_analytical_correlation_function(L, htf, eb, J, sites, T, steps, mode='grid cell state', ax=ax)
ax.plot([45000000,45000000], [0,0.25], '--k', label='10 minute mark')
ax.plot([tau_occupancy,tau_occupancy], [0,0.25], ':r', label='correlation time mark')
xs = np.linspace(0,T,steps+1)
ax.plot(xs, sigma_squared_occupancy*np.exp(-xs/tau_occupancy), ':')
ax.legend(['occupancy', 'grid cell state', '10 minute mark', 'correlation time mark', 'exponential'])
```

We see that for small times, the correlation of the grid cell state is higher than that of the occupancy. This is because the grid cell state has a higher variane around the mean in the steady state (t=0).

It is thus important that we use the correct function.

# Single grid-cell model

```python
from MarkovChainClass import FullRates, FullEnergies, FullMatrix

beta  = 1
htf   = -15
eb    = 11
J     = 4
L     = 4
sites = np.arange(L)

T = 45000000

mode = 'standard'
periodic = True

R = FullRates(mode, htf=htf, eb=eb, J=J)
E = FullEnergies(mode, htf, eb, J)
M = FullMatrix(L, R.rate_function, E.energy_function, sites, periodic)
mat = M.create_Matrix(sparse=False)
occupancies = M.states.calculate_state_occupancies(sites)
grid_cell_states = M.int_states.T
P_steady = M.calculate_steady_state()

Expected_kon  = np.exp(beta*htf)
Expected_koff = 0

for state, probability in zip(grid_cell_states, P_steady):
    N = state[1]+state[-1]
    Expected_koff += (np.exp(-beta*(eb+N*J)))*probability

tau = 1/(Expected_kon+Expected_koff)
mu = grid_cell_states[0]@P_steady
sigma_squared = (grid_cell_states[0]-mu)**2@P_steady
    
print('k_on: ', Expected_kon)
print('k_off:', Expected_koff, f'(min: {np.exp(-beta*eb)}, max: {np.exp(-beta*(eb+2*J))})')
print('tau:', tau)
print('variance:', 2*sigma_squared*tau/T)
```

Taking the average rate changes the dynamics of the grid cell state. If in most cases the on state is short-lived, but in some cases it is very long lived, then the average rate will be somewhere in between. Rather than getting many short-lived events and some very long lived ones we then get many events of intermediate length.

It is not possible to simply ingnore the coupling between grid cells.
