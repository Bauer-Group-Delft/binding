# Experiment 1: Timetraces

## Experiment 1.1: Timetrace of TF binding

Timetrace of the number or fraction of binding sites occupied in a single simulation.

```python
from Tests import timetrace

## -------------------------------- Define parameters --------------------------------

htf = -16
eb  = 10.5
J   = 5.5
T   = 1000*22500000

## -------------------------------- Define binding sites --------------------------------

# All sites are binding sites
L   = 6
sites = np.arange(L)

# Array of binding sites flanked by nonbinding sites
# NSites = 6
# padding = 10
# L = NSites+2*padding
# sites = np.arange(padding, NSites+padding)

## -------------------------------- Create timetrace --------------------------------

datafile = timetrace(TF_Grid_1D, T, htf, eb, L=L, J=J, sites=sites, mode='avgBinding')
```

Results:
- For J=6, eb=0.9, htf=-7, L=10 (all binding sites) two clearly distinguishable steady states (~0 and ~1) are observed. The system randomly switches between these states, leading to a wide range of mean occupancies for different runs. (Though this range should decrease as the run time increases due to longer time averaging.)
- For a larger number of (binding) sites (e.g. L=100), these states are less clearly visible, since it takes time for the change in state to spread along the array of binding sites. Therefore, when there are more binding sites, the site will be less likely to all be in the same state and averaging over the bidning sites is more likely to lead to intermediate occupancies.
- For a small (N=6) array of binding sites surrounded by nonbinding sites, the two states are observed, but the on-state (~1) is less likely than when all sites are binding sites. This is because the non-binding sites are less likely to be filled.

### Running average

Running average over a timetrace of transcription factor binding. The running average is calculated by taking the average over a time window of width T_avg, which is shifted along the signal. The start position of the window is indicated on the x-axis.

```python
from Tests import running_avg, plot_gillespie

## -------------------------------- Define parameters --------------------------------

T_avg = 10000
dt    = T/1000

## -------------------------------- Calculate running average --------------------------------

t, NBound = datafile.get_NBoundPerSite()
timepoints, running_avg = running_avg(t, NBound, T_avg, dt)

## -------------------------------- Plot running average --------------------------------

plot_gillespie(datafile.convert_time_to_seconds(t), NBound)
plt.plot(datafile.convert_time_to_seconds(timepoints), running_avg, label='running average,\n$T_{avg} = $'+f'${T_avg}$, $dt = {dt}$')
plt.xlabel('time, $t$')
plt.ylabel('number of bound molecules')
plt.legend()
```

## Experiment 1.2: Distribution of number of bound TFs

### Single grid cell

We can compare a 1D grid of length 1 to a 0D grid to see the effect of clustering and periodic boundaries.

```python
from Experiments import binding_distribution_timetrace
from MarkovChain import create_Matrix, solve_MasterEquation, compact_v
from Grid0D import TF_Grid_0D

## -------------------------------- Define parameters --------------------------------

beta  = 1
htf   = -1
eb    = 1
J     = 2.6

L     = 1
sites = np.arange(L)

T     = 20
runs  = 1000
steps = 1000

## -------------------------------- Simulation --------------------------------

plt.figure()

SimulationGrid = TF_Grid_1D(L, htf, eb, J, sites)
datafile = DataFile(None)
Simulation = GillespieSim(SimulationGrid, datafile)

_, _ = binding_distribution_timetrace(Simulation, datafile, len(sites), runs, T, steps)

plt.gca().set_prop_cycle(None)

# 0D grid for comparison
SimulationGrid = TF_Grid_0D(htf, eb+2*J)
datafile = DataFile(None)
Simulation = GillespieSim(SimulationGrid, datafile)

_, _ = binding_distribution_timetrace(Simulation, datafile, len(sites), runs, T, steps, linestyle=':')

plt.legend(['1D: 0 TFs bound', '1D: 1 TFs bound', '0D: 0 TFs bound', '0D: 1 TFs bound'])
```

Note that 0D grid starts off with a certain fill fraction rahter than a fixed number of TFs. Despite this, we can still compare the steady states. We see that in the 1D grid there are much more TFs bound in the steady state. This is because the bound TF interacts with itself via the periodic boudnary conditions. This is thus an artifact caused by the clustering in combination with the periodic BCs. If we add 2J to eb in the 0D model, it also goes to the steady state predicted by the master equation model.
