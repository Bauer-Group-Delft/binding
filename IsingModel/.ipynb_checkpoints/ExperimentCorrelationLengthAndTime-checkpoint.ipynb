# Correlation length

```python
from Tests import simulate, time_mean
from Grid1D import TF_Grid_1D
from Experiments import remove_transient

## Spatial correlation
def calc_spatial_correlation(htf, eb, J, L, sites, T, T_trans):
    datafile = simulate(TF_Grid_1D, T, htf, eb, L=L, J=J, sites=sites)
    t, grids = datafile.get_grids() 
    
    # Remove transient
    t, grids = remove_transient(t, T_trans, grids)

    # Calculate correlation
    grids[np.where(grids==0)] = -1
    mean = time_mean(t, grids)
    
    cor_mean = mean*mean[int(L/2)] # If J=0, the mean is 0.
    correlation_zero_mean = time_mean(t, np.array([grid*grid[int(L/2)] for grid in grids]))
    correlation = correlation_zero_mean - cor_mean
    cor_clustering = correlation/(1-cor_mean)
    
    return correlation, correlation_zero_mean, cor_mean, cor_clustering

def plot_spatial_correlation(htf_list, eb, J, L, sites, T, T_trans):
    fig, ax = plt.subplots(1,2,figsize=(12,4))

    for htf in htf_list:
        print(f'htf = {htf}')

        correlation, correlation_zero_mean, cor_mean, cor_clustering = calc_spatial_correlation(htf, eb, J, L, sites, T, T_trans)

        # Plot correlation
        ax[0].plot(np.arange(L), correlation, label = '$h_{tf}$ = '+f'{htf}')
        ax[1].plot(np.arange(L), cor_clustering, label = '$h_{tf}$ = '+f'{htf}')

        clear_output(wait=True)

    beta = 1
    x = np.linspace(0,L-1,1000)
    xi = -1/np.log(np.tanh(beta*J/4))
    ax[0].plot(x, np.exp(-np.abs(x-int(L/2))/xi), ':k', label = 'Ising model, J/4')
    ax[1].plot(x, np.exp(-np.abs(x-int(L/2))/xi), ':k', label = 'Ising model, J/4')

    ax[0].set_title('Correlation')
    ax[0].set_xlabel('Sites')
    ax[0].set_ylabel('Correlation')
    ax[0].legend()
    
    ax[1].set_title('Rescaled correlation')
    ax[1].set_xlabel('Sites')
    ax[1].set_ylabel('Correlation')
    ax[1].legend()
    
    return fig, ax
```

```python
beta = 1
eb = 0
J = 2.6
htf_list = [-1.7, -J, -3.5]
L = 50
sites = np.array([])
T_trans = 1000
T = 100*T_trans

for htf in htf_list:
    print(f'htf = {htf}')
    
    correlation, correlation_zero_mean, cor_mean, cor_clustering = calc_spatial_correlation(htf, eb, J, L, sites, T, T_trans)

    plt.figure()
    plt.title('$h_{tf}$ = '+f'{htf}')
    plt.plot(np.arange(L), correlation, label = 'correlation with\ncalculated mean')
    plt.plot(np.arange(L), correlation_zero_mean, label = 'correlation with\nmean 0')
    plt.plot(np.arange(L), cor_mean, label = 'correlation due\nto mean')
    plt.plot(np.arange(L), cor_clustering, label = 'correlation due\nto clustering')
    x = np.linspace(0,L-1,1000)
    xi = -1/np.log(np.tanh(beta*J/4))
    plt.plot(x, np.exp(-np.abs(x-int(L/2))/xi), ':k', label = 'Ising model, J/4')
    plt.xlabel('Site')
    plt.ylabel('Correlation')
    plt.legend()
    
    clear_output(wait=True)
```

When compensating the effect of $J$ on the mean by using a negative $h_{th}$, the correlation length in our model matches the correlation length of the Ising model with $J_{Ising}=J/4$.

Note: when $J$ is set too high, $h_{tf}=-J$ becomes really small, therefore it takes very long before the grid fills up. As a result, a long runtime is required to obtain the intended results.

Because a higher $J$ makes the existence of clusters more likely over the whole field, it becomes more difficult to determine where one cluster ends and another begins. Clusters essentially merge together to cover the whole field. This behaviour is different form the Ising model, and therefore we are likely not able to predict it using the Ising model.
Here, we only see the effect of $J$ on the correlation and not on the mean.

The more $h_{tf}$ deviates from $-J$, the higher the correlation calculated with mean 0 (orange line) will be in general, since the mean is not zero. This makes it harder to compare the system results to the ising model. However, we might be able to find a rescaling such that the models still compare, e.g. mean correlation + (1 - mean correlation)\*exponential decay. If this is indeed possible, it suggests that the correlation between sites caused by coupling and the correlation between sites caused by the mean can be separated. The correlation caused by coupling is then expected to match that of the Ising model.

Using this method, we find a smaller correlation length that that of the Ising model when htf is not equal to -J. This suggests that the correlation due to clustering is not independent of the correlation due to the mean. This is likely because not clustering not only affects the correlation, but also the mean. Therefore, when subtracting the product of the means, part of the effect of $J$ is removed from the correlation.

If the mean is larger than zero (i.e. being bound is more likely than not being bound), there is already an increased probability that two neighboring sites are in the same state ($p^2+(1-p)^2$ is maximized when $p=0$ or $p=1$ and minimized when $p=1/2$). Therefore, clustering can not increase this probability as much anymore, leading to a correlation with the site itself that is less than one.

If the mean is less than zero, the probability of sites being filled is decreased, and since clustering only occurs between filled sites, so is the effect of clustering.

## Effect of $h_{tf}$

```python
htf_list = np.linspace(-10,-1,10)
eb = 0
J = 3
L = 50
sites = np.array([])
T_trans = 1000
T = 100*T_trans

fig, ax = plot_spatial_correlation(htf_list, eb, J, L, sites, T, T_trans)
```

The correlation length is maximizxed when $h_{tf}=-J$ (when eb=0). If $h_{tf}$ is lower, the probability of clusters occuring is lower, and clusters won't spread as far/won't grow as big. If $h_{tf}$ is higher, the probability of sites being filled is already increased, and clustering can not increase it as much anymore.

Thus, there is an optimal clustering strength at which deviations form the mean propagate further through the grid. I expect this to corresponds to the parameter values for which the system has most memory, i.e. longest equilibration time.

## Effect of $e_b$

```python
eb_list = [1,3,6]
J = 2
L = 50
sites = np.arange(L)
T_trans = 1000
T = 100*T_trans

for eb in eb_list:
    htf_list = [-eb-J-2, -eb-J-1, -eb-J, -eb-J+1, -eb-J+2]
   
    fig, ax = plot_spatial_correlation(htf_list, eb, J, L, sites, T, T_trans)

    ax[0].set_title('correlation, $e_b$ = '+f'{eb}')
    ax[1].set_title('rescaled correlation, $e_b$ = '+f'{eb}')
```

As expected, when all sites are equal, $e_b$ caues a shift in $h_{tf}$. For higher $e_b$ there seems to be more variation, as we also observed in the occupancy against $h_{tf}$ curves.

$e_b$ thus causes an increase in local concentration at binding site arrays, causing clustering to have a different effect there than in reagions without binding sites. This means that for low concentations, custers can still form and have a large effect at binding site arrays, while no clusters are present in the rest of the cell. At high concentrations however, the binding sites would already be occupied and clustering would not affect them much, but would have more effect in the rest of the cell.

Obviously it will also be important to take into account depletion effects when considering large grids with many binding sites. This will be examined in another experiment.

## The effect of binding sites

*Binding site array on one half of the grid:*

```python
htf_list = np.linspace(-10,-1,10)
eb = 2
J = 3
L = 50
sites_list = [np.arange(int(L/2),L), np.arange(int(L/2)+1,L)]
T_trans = 1000
T = 100*T_trans

for i,sites in enumerate(sites_list):
    fig, ax = plot_spatial_correlation(htf_list, eb, J, L, sites, T, T_trans)
    
    if i==0:
        text = 'binding site'
    elif i==1:
        text = 'non-binding site'
    
    ax[0].set_title(f'correlation, {text}')
    ax[1].set_title(f'rescaled correlation, {text}')
```

As expected, calculating the correlation to a side on the edge of a binding site array, we see that the correlation length in the direction of the non-binding sites is longest when $h_{tf}=-J$ and the correlation length in the direction of the bindign sites is longest when $h_{tf}=-eb-J$. Thus, the larger $e_b$, the larger the difference correlation length could be between bindign sites and non-bindign sites.

Note that if the site compared to which we calculate the correlation is a binding site, the maximal correlation in the direction of the binding sites comes close to the prediction of the Ising model, while if the site is a non-binding site, the maximal correlation on the non-binding site comes close to this prediction. This suggests that the correlation lenght is highest along sites with the same $e_b$.

*Single binding site in the center:*

```python
htf_list = np.linspace(-10,-1,10)
eb = 5
J = 3
L = 50
sites_list = [[], [int(L/2)]]
T_trans = 1000
T = 100*T_trans

for sites in sites_list:
    fig, ax = plot_spatial_correlation(htf_list, eb, J, L, sites, T, T_trans)
        
    ax[0].set_title(f'correlation, {len(sites)} binding sites')
    ax[1].set_title(f'rescaled correlation, {len(sites)} binding sites')
```

The corrlation length is still largerst for $h_{tf}=-J$, but this is no longer the highest peak.

$e_b=2$, $J=3$: The correlation length does seem to be affected very much by the presence of the bindign site.

$e_b=5$, $J=3$: The correlation length is clearly shorter.

Why is the concentration for best propagation through non-binding sites not the optimal concentration here? This indeed seems to be the one for which the correlation decays slowest, but the autocorrelation of the binding site is very low (since it would have a very high mean occupancy). When rescaling the gray line indeed seems one of the best, but it is not significantly bettwe then some of the others. This is likely because the high mean occupancy of the binding site is also propagated through the grid at those concentreations, while only the deviations from this mean are relevant for the correlation. 

*Array of binding sites at the center:*

```python
htf_list = np.linspace(-10,-1,10)
eb = 2
J = 5
L = 50
sites = np.arange(23,28)
T_trans = 10000
T = 100*T_trans

plot_spatial_correlation(htf_list, eb, J, L, sites, T, T_trans)
```

We expect to see too scales in the correlation, slow decay along the binding sites and faster decay when the non-binding sites are reached. This effect is ecpected to be more extreme when $e_b$ is higher.

There indeed seem to be two scales, where the second one is faster (at least for the curve with higherst correlation). We see that for high $e_b$ the correlation is relatively high along the binding sites, and then rapidly drops along the non-binding sites, because at the cocnentration/$h_{tf}$ for which the correlation along the binding sites is maximized, the on rate is too low to propagate the infromation/clusters along the non-binding sites. This again shows that $e_b$ causes a different effective concentration, therefore when the cocentration is optimal for propagation of information along binding sites, it is not optimal for propagation of infromation along non-binding sites.

Note that the $h_{tf}$ at which the correlation length is maximized is lower now, just as we saw that the shift of the occupancy curve is lower when there is a padding of nonbinding sites.

Overall it seems that the correlation length will always be less or equal than the prediction from the Ising model, which is the correlation distance when $h_{tf}=-J-eb$, which as we have seen before is the maximum correlation distange than can be achieved.

If there was actual cooperativity the effect is expected to decay much faster with distance between bindign sites, or even disappear immediately once the distance becomes too large to bridge.

# Correlation time

```python
def calc_temporal_correlation(htf, eb, J, L, sites, T, T_trans):
    datafile = simulate(TF_Grid_1D, T, htf, eb, L=L, J=J, sites=sites)
    t_out, grids_out = datafile.get_grids() 
    
    # Remove transient
    t_out, grids_out = remove_transient(t_out, T_trans, grids_out, True)
    
    # Time sampling
    t, grids = datafile.sample_output(t_out, grids_out, T-T_trans, 1000)

    # Calculate correlation
    grids[np.where(grids==0)] = -1
    mean = np.mean(grids, axis=1)
    
    cor_mean = mean*mean[0]
    correlation_zero_mean = np.mean([grid*grids[0] for grid in grids], axis=1)
    correlation = correlation_zero_mean - cor_mean
    cor_clustering = correlation/(1-cor_mean)
    
    return datafile.convert_time_to_seconds(t), correlation, cor_clustering

def plot_temporal_correlation(htf_list, eb, J, L, sites, T, T_trans):
    fig, ax = plt.subplots(1,2,figsize=(12,4))

    for htf in htf_list:
        print(f'htf = {htf}')
        
        t, correlation, cor_clustering = calc_temporal_correlation(htf, eb, J, L, sites, T, T_trans)
        
        ax[0].plot(t, correlation, label = '$h_{tf}$ = '+f'{htf}')  
        ax[1].plot(t, cor_clustering, label = '$h_{tf}$ = '+f'{htf}')  
        
        clear_output(wait=True)
    
    ax[0].set_title('Correlation')
    ax[0].set_xlabel('Time, $t$ (seconds)')
    ax[0].set_ylabel('Correlation')
    ax[0].legend()
    
    ax[1].set_title('Rescaled correlation')
    ax[1].set_xlabel('Time, $t$ (seconds)')
    ax[1].set_ylabel('Correlation')
    ax[1].legend()
    
    return fig, ax
```

```python
eb_list = [0,1,3,6]
J = 3
L = 1000
sites = np.arange(L)
T_trans = 100
T = 300

for eb in eb_list:
    htf_list = [-eb-J-2, -eb-J-1, -eb-J, -eb-J+1, -eb-J+2]
    
    fig, ax = plot_temporal_correlation(htf_list, eb, J, L, sites, T, T_trans)
    
    ax[0].set_title(f'Time correlation, $e_b = {eb}$')
    ax[1].set_title(f'Rescaled time correlation, $e_b = {eb}$')
```

Again we see that as $e_b$ becomes larger, the $h_{tf}$ at which the correlation time is larger deviates more from the expected value of $-(e_b+J)$.

# Correlation function

```python
## Temporal correlation
from Predictions import predict_time_correlation
from Tests import simulate
from Experiments import remove_transient
from Grid1D import TF_Grid_1D

## Correlation function
def correlation_function(signal, end):
    mean = np.mean(signal)
    correlation = np.zeros(end+1)
    correlation[0] = np.mean((signal-mean)**2)
    for i in range(1,len(signal)):
        if i>end:
            break
        correlation[i] = np.mean((signal[:-i]-mean)*(signal[i:]-mean))
    return correlation

def calc_correlation_function(htf, eb, J, L, sites, T, T_trans=0, T_stop=50, sample_freq=1):
    datafile = simulate(TF_Grid_1D, T, htf, eb, L=L, J=J, sites=sites)
    t_output, output = datafile.get_grids()
    
    # remove transient
    t_output, output = remove_transient(t_output, T_trans, output, True)
    
    # Sample data
    steps = sample_freq*T
    stop  = sample_freq*T_stop
    times, grids = datafile.sample_output(t_output, output, T, steps)
    
    # Calculate mean and standard deviation over sites
    mean_correlation = np.mean([correlation_function(grids[:,cell], stop) for cell in range(len(grids[0]))], axis=0)
    std_correlation  = np.std([correlation_function(grids[:,cell], stop) for cell in range(len(grids[0]))], axis=0)
    
    # Calculate prediction in the absence of clustering
    predicted_correlation = predict_time_correlation(times[:stop+1], htf, eb) 

    return datafile.convert_time_to_seconds(times[:stop+1]), mean_correlation, std_correlation, predicted_correlation

def plot_correlation_function(htf_list, eb, J, L, sites, T, T_trans=0, T_stop=50, sample_freq=1, prediction=False):
    fig, ax = plt.subplots(1,2,figsize=(12,4))
    
    for i, htf in enumerate(htf_list):
        print(f'htf = {htf}')
        
        times, mean_correlation, std_correlation, predicted_correlation = calc_correlation_function(htf, eb, J, L, sites, T, T_trans, T_stop, sample_freq)
        
        ax[0].plot(times, mean_correlation, c=f'C{i}', label = '$h_{tf}$ = '+f'{htf}')
        ax[0].fill_between(times, mean_correlation-std_correlation, mean_correlation+std_correlation, color=f'C{i}', alpha=0.3)
        ax[1].plot(times, mean_correlation/mean_correlation[0], c=f'C{i}', label = '$h_{tf}$ = '+f'{htf}')
        ax[1].fill_between(times, (mean_correlation-std_correlation)/mean_correlation[0], (mean_correlation+std_correlation)/mean_correlation[0], color=f'C{i}', alpha=0.3)
        
        if prediction:
            ax[0].plot(times, predicted_correlation, ':k')
            ax[1].plot(times, predicted_correlation/predicted_correlation[0], ':k')
        
        clear_output(wait=True)
    
    ax[0].set_xlim(0,times[-1])
    ax[0].set_title('Correlation')
    ax[0].set_xlabel('Time, t')
    ax[0].set_ylabel('Correlation, C(t)')
    ax[0].legend()
    
    ax[1].set_xlim(0,times[-1])
    ax[1].set_title('Rescaled correlation')
    ax[1].set_xlabel('Time, t')
    ax[1].set_ylabel('Correlation, C(t)')
    ax[1].legend()
    
    return fig, ax
```

```python
from Tests import simulate

## -------------------------------- Define parameters --------------------------------

htf_list = [-5]
eb  = 1
L   = 10
J   = 0
T   = 20000
sites = np.arange(L)

plot_correlation_function(htf_list, eb, J, L, sites, T, T_stop=50, prediction=True)
```

In the absence of clustering, the correlation time matches the expectation.

## Correlation time of occupancy

```python
## Correlation function for occupancy
def calc_occupancy_correlation_function(htf, eb, J, L, sites, T, T_trans=0, T_stop=50, sample_freq=1):
    datafile = simulate(TF_Grid_1D, T, htf, eb, L=L, J=J, sites=sites)
    t_output, output = datafile.get_grids()
    
    # remove transient
    t_output, output = remove_transient(t_output, T_trans, output, True)
    
    # Sample data
    steps = sample_freq*T
    stop  = sample_freq*T_stop
    times, grids = datafile.sample_output(t_output, output, T, steps)
    
    # Calculate mean and standard deviation over sites
    occupancies = np.sum([grids[:,i] for i in sites], axis=0)/len(sites)
    correlation = correlation_function(occupancies, stop)

    return datafile.convert_time_to_seconds(times[:stop+1]), correlation
```

```python
from MarkovChain import plot_analytical_correlation_function

beta = 1
htf = -2
eb  = 1
L   = 6
J   = 3
T   = 200000
T_trans = 2000
T_stop = 100
sites = np.arange(L)

fig, ax = plt.subplots()
times = np.linspace(0,T_stop,T_stop+1)

correlations = []
for i in range(10):
    _, correlation = calc_occupancy_correlation_function(htf, eb, J, L, sites, T, T_trans=T_trans, T_stop=T_stop)
    correlations.append(correlation)
    clear_output(wait=True)

mean_correlation = np.mean(correlations, axis=0)
std_correlation  = np.std(correlations, axis=0)
plt.plot(times, mean_correlation, color='C1', label='simulation results')
plt.fill_between(times, mean_correlation-std_correlation, mean_correlation+std_correlation, color='C1', alpha=0.3)
    
plot_analytical_correlation_function(times, beta, htf, eb, J, L, sites, ax=ax)
plt.legend()
```

# Memory

Use correlation between different sites and times to determine the evolution of correlation.

```python
## Spatial and temporal correlation
def calc_full_grid_correlation(htf, eb, J, L, sites, T, T_trans, runs, steps):
    # Simulate a grid over time, repeat for many runs.
    grids = np.zeros((runs, steps+1, L))
    for run in range(runs):
        print(f'J = {J}, run {run}')
        
        datafile = simulate(TF_Grid_1D, T, htf, eb, J=J, L=L, sites=sites)
        t_out, grids_out = datafile.get_grids() 

        # Remove transient
        t_out, grids_out = remove_transient(t_out, T_trans, grids_out, True)

        # Time sampling
        t, grids[run] = datafile.sample_output(t_out, grids_out, T-T_trans, steps)
        clear_output(wait=True)

    # Calculate correlation between center grid cell at t=0 and all other grid cells at all times.
    grids[np.where(grids==0)] = -1
    mean = np.mean(grids, axis=0)

    cor_mean = mean*mean[0][int(L/2)]
    correlation_zero_mean = np.mean([grid*grid[0][int(L/2)] for grid in grids], axis=0)
    correlation = correlation_zero_mean - cor_mean

    return datafile.convert_time_to_seconds(t), correlation

def plot_full_grid_correlation(htf, eb, J_list, L, sites, T, T_trans, runs, steps, rescaled=True):
    for J in J_list:
        times, correlation = calc_full_grid_correlation(htf, eb, J, L, sites, T, T_trans, runs, steps)

        plt.figure()

        if rescaled:
            plt.imshow(correlation.T/correlation[0][int(L/2)], aspect=20, vmin=0, vmax=1)
            plt.title(f'Rescaled correlation, J = {J}.')        
        else:
            plt.imshow(correlation.T, aspect=20, vmin=0, vmax=1)
            plt.title(f'Correlation, J = {J}.')

        plt.xlabel('Time, $t$ (seconds)')
        plt.ylabel('Site')
        tick_interval = 200
        plt.xticks(np.arange(0,steps+1,tick_interval), ['{:.2e}'.format(i) for i in times[::tick_interval]])
        
```

```python
htf = -3
eb = 0
J_list = [2, 3, 7]
L = 20
sites = np.arange(L)
T_trans = 100
T = T_trans + 100
runs = 500
steps = 1000

plot_full_grid_correlation(htf, eb, J_list, L, sites, T, T_trans, runs, steps)
        
```
