# Experiment 1: Effect of binding site and clustering on occpancy

Calculate the mean and standard deviation of the mean occupancy (mean over time) of the different sites over a number of runs. Mean and standard deviation of occupancy are plotted as lines against the index of the site. Three scenarios are examined, one without binding site and clustering, one with bindign site without clustering and one with binding site and clustering. This way the effect of the binding site and the clustering on the mean occupation of a site can be determined.

```python
from Tests import plot_grid
from Experiments import calculate_occupancies, plot_occupancy_against_site

L     = 50
htf   = -3
T     = 20000
sites = np.array([24])
runs  = 5

labels = []
occupancies_list = []
params = [(0,0), (0,2.6), (1,0), (1,2.6)]
for eb, J in params:
    labels.append(f'$e_b={eb}$, $J={J}$')
    occupancies_list.append(calculate_occupancies(TF_Grid_1D, runs, T, htf, eb, L=L, J=J, sites=sites))
    clear_output(wait=True)
    
# Plot results
fig, axs = plt.subplots(4,1)
fig.tight_layout(pad=3)

mean_occupancies_list = np.mean(occupancies_list, axis=1)
for i,occupancies in enumerate(occupancies_list):
    plot_grid(np.mean(occupancies, axis=0), (np.min(mean_occupancies_list), np.max(mean_occupancies_list)), fig, axs[i])
    axs[i].set_title(labels[i])
    axs[i].set_yticks([])
    
linestyles = len(params)*['-']
colors = [f'C{i}' for i in range(len(params))]
plot_occupancy_against_site(occupancies_list, labels, colors, linestyles)
```

Note that clustering overall increases the occupancy, but it most strongly increases the occupancy close to the binding site, since there molecules bind for longer. This suggests that for this set of parameters (htf = -3, eb = 0.9, J =2.6), clusters will moslty form around the binding site. (Check this quantitatively.)

We also observe an increase in standard deviation caused by clustering.

## Effect of $J$ on mean occupacy of grid with all binding sites

Use the steady state solution to the Master equation to predict the mean occupancy of an grid with only binding sites in the presence of clustering.

```python
from MarkovChain import calculate_steady_state

beta = 1
htf  = -3
eb   = 1
J_list = np.linspace(0,10,21)

L   = 4
sites = np.arange(L)

occupancy = np.zeros(len(J_list))
for j,J in enumerate(J_list):
    states, int_states, steady = calculate_steady_state(L, beta, htf, eb, J, sites)
    occupancy[j] = np.sum(np.sum(int_states, axis=1)*steady)/L
    
plt.plot(J_list, occupancy)
plt.ylim([0,1.01])
plt.xlabel('$J$')
plt.ylabel('occupancy')
```

At low $J$, clustering is not strong enough to make the clusters last long enough to grow. When $J$ becomes large enough, the clusters remain long enough to grow and fill up most of the grid. Making $J$ even larger will ake the clusters last longer still, but this does not matter as much since the cluster(s) already take up most of the grid. Therefore increasing $J$ more does not improve cluster growth, but only decreases the probability of molecules in the cluster unbinding and having to rebind. So there is only a sall range of $J$ values for which the clusters actually last long enough and have enough space to grow. The graph seems to be steepest around the point were $p=0.5$ and $h_{tf}+e_b=-J$.

```python
from MarkovChain import calculate_steady_state

beta = 1
htf  = -5
J_list = np.linspace(0,10,21)

L   = 4
sites = np.arange(L)

for eb in range(10):
    occupancy = np.zeros(len(J_list))
    for j,J in enumerate(J_list):
        states, int_states, steady = calculate_steady_state(L, beta, htf, eb, J, sites)
        occupancy[j] = np.sum(np.sum(int_states, axis=1)*steady)/L

    plt.plot(J_list+htf+eb, occupancy)
    plt.ylim([0,1.01])
    plt.xlabel('$J+(h_{tf}+e_b)$')
    plt.ylabel('occupancy')
```

The curve becomes less steep as $e_b$ increases, because a higher $e_b$ also means that the clusters remain for longer. And thus a smaller $J$ can then already cause a cluster to grow. 

```python
from MarkovChain import calculate_steady_state

beta = 1
eb  = 3
J_list = np.linspace(0,10,21)

L   = 4
sites = np.arange(L)

for htf in np.arange(-10,0):
    occupancy = np.zeros(len(J_list))
    for j,J in enumerate(J_list):
        states, int_states, steady = calculate_steady_state(L, beta, htf, eb, J, sites)
        occupancy[j] = np.sum(np.sum(int_states, axis=1)*steady)/L

    plt.plot(J_list+htf+eb, occupancy)
    plt.ylim([0,1.01])
    plt.xlabel('$J+(h_{tf}+e_b)$')
    plt.ylabel('occupancy')
```

Similarly, the curve becomes less steep as $h_{tf}$ increases, because TFs will appear faster and allow clusters to grow faster.

```python
from MarkovChain import calculate_steady_state

beta = 1
J_list = np.linspace(0,10,41)

L   = 4
sites = np.arange(L)

for htf in np.arange(-10,0):
    eb = -5-htf
    occupancy = np.zeros(len(J_list))
    for j,J in enumerate(J_list):
        states, int_states, steady = calculate_steady_state(L, beta, htf, eb, J, sites)
        occupancy[j] = np.sum(np.sum(int_states, axis=1)*steady)/L

    plt.plot(J_list, occupancy)
    plt.ylim([0,1.01])
    plt.xlabel('$J$')
    plt.ylabel('occupancy')
```

As long as $h_{tf}+e_b$ remains the same, the curve is the same. So $h_{tf}+e_b$ detemines both the position and the steepness of the curve. So it does not matter whether TFs appear faster and remmain shorter or appear slower and remain longer. At least not for the mean occupancy. How does it affect the correlation length?

## Binding site array with and without clustering

```python
from Experiments import plot_kymograph_1D

htf = -13
eb  = 10.5
J_list   = [0,4.5]
L   = 26
sites = np.arange(10,16)
T   = 22500000

for J in J_list:
    plot_kymograph_1D(htf, eb, J, L, sites, T, steps = 1000)
```

This simulation proves that with the right combination of parameters (eb=2,J=2,htf=-3.5), we can get lots of TF binding to the bidning sites and relatively little binding to other sites. Next we need to check how this influences the standard deviation of the binding site occupancy.

For parameters (eb=4,J=2.6,htf=-4) we see that clustering almost doubles the mean binding site occupancy and thereby decreases the standard deviation within the binding site array. The standard deviation at the edges of the array howerver, does not seem to be significantly decrease. This is because clustering makes the change in mean occupancy between binding and non-bindign sites more gradual, leading to fluctuations at the boundary.

Note that this simulation considers the standard deviation over a single run, while we are really interested in the standard deviation of the mean occupation over different runs. This will depend on the correlation times, which will likely be different for the system with and the system without clustering.

# Experiment 2: Effect of clustering on occupancy for different number of binding sites

## Experiment 2.1: Array of binding sites

```python
# Effect of multiple binding sites on occupancy

from Experiments import calculate_occupancies, plot_occupancy_against_site

htf   = -3
eb    = 1
T     = 20000
runs  = 50

L_list = np.arange(1,10)
J_list = [0,2.6]

occupancies_list = []
for L in L_list:
    for J in J_list:
        print(f'L = {L}, J = {J}.')
        sites = np.arange(L)
        occupancies_list.append(calculate_occupancies(TF_Grid_1D, runs, T, htf, eb, L=L, J=J, sites=sites))
        clear_output(wait=True)

# Plot results
labels = [f'L = {i}, J = {j}.' for i in L_list for j in J_list]
linestyles = ['-', '-.']*int(len(occupancies_list)/2)
colors = [f'C{int(i/2)}' for i in range(len(occupancies_list))]
plot_occupancy_against_site(occupancies_list, labels, colors, linestyles)
```

Having multiple sites should lead to averaging, therefore decreasing the standard deviation even without cooperativity. 

The mean occupancy is higher with clustering as expected. In the absense of clustering the mean occupancy does not depend on the number of binding sites, but in the presence of clustering it does. It seems like more binding sites leads to higher occupancy, but the increase of occupancy achieved by adding a binding site decreases with the number of binding sites. This makes sense, since the the more binding sites are present, the more interactions are possible. However, binding sites are added at the ends and for longer arrays of binding sites, the ends have less and less influence on the overall mean occupancy, since the number of ends is always 2 while the number of binding sites increases. So the effect of adding a binding site at the ends decreases too.

Note that the occupancy for L=1 is very high due to the periodic boundary conditions. This means that when a TF is present it always clusters.

```python
# Mean and standard deviation of occupancy against 

mean_occupancies = np.array([[np.mean(occupancies_list[2*i+j]) for j in range(len(J_list))] for i in range(len(L_list))])
std_occupancies  = np.array([[np.std(occupancies_list[2*i+j]) for j in range(len(J_list))] for i in range(len(L_list))])
# Note: mean and standard deviation are over sites and over runs simultaneously.

fig, ax = plt.subplots(1,2, figsize=(10,4))
fig.tight_layout(pad=4)

ax[0].errorbar(L_list, mean_occupancies[:,0], yerr=std_occupancies[:,0], capsize=5)
ax[0].errorbar(L_list, mean_occupancies[:,1], yerr=std_occupancies[:,1], capsize=5)
ax[0].set_xlabel('number of binding sites')
ax[0].set_ylabel('mean occupancy')
ax[0].legend(['$J=0$','$J=2.6$'])

ax[1].plot(L_list, std_occupancies, '-.')
ax[1].set_xlabel('number of binding sites')
ax[1].set_ylabel('standard deviation of occupancy')
ax[1].legend(['$J=0$','$J=2.6$'])
```

There should not be a difference in mean occupancy with and without clustering if there is only one binding site. Yet a difference is consistently observed. This is caused by the periodic boundary condition. Since a sigle site would be it's own neighbor, thereby strengthening its own interaction. To prevent this, we add a number of nonbinding sites on each side of the binding site array. We use the correlation length found in experiment 5 to determine the nubmer of sites that should be added on each side.

## Experiment 2.2: Array of binding sites with non-binding sites at the ends

```python
# Effect of multiple binding sites on occupancy

from Experiments import calculate_occupancies, plot_occupancy_against_site
import math as m

beta = 1
htf  = -5
eb   = 5
T    = 20000
runs = 50

L_list = np.arange(1,11)
J_list = [0,2.6]
padding = m.ceil(5*(-1/np.log(np.tanh(beta*J_list[-1]/4))))
print(f'Adding a padding of {padding} sites.')

occupancies_list = []
occupancies_list2 = []
bs_occupancies_list = []
for L in L_list:
    for J in J_list:
        print(f'L = {L}, J = {J}.')
        sites = np.arange(padding,L+padding)
        occupancies, bs_occupancies = calculate_occupancies(TF_Grid_1D, runs, T, htf, eb, L=L+2*padding, J=J, sites=sites, mode='all')
        occupancies_list.append(occupancies) # store data of all sites
        occupancies_list2.append(np.array(occupancies)[:,padding:-padding]) # only store data of binding sites
        bs_occupancies_list.append(bs_occupancies)
        clear_output(wait=True)

# Plot results
print(f'Paddding: {padding} sites.')
labels = [f'L = {i}, J = {j}.' for i in L_list for j in J_list]
linestyles = ['-', '-.']*int(len(occupancies_list)/2)
colors = [f'C{int(i/2)}' for i in range(len(occupancies_list))]
plot_occupancy_against_site(occupancies_list, labels, colors, linestyles)
plot_occupancy_against_site(occupancies_list2, labels, colors, linestyles)
```

```python
bs_occupancies_list = np.reshape(bs_occupancies_list, (len(L_list),len(J_list),runs))
mean_occupancies = np.mean(bs_occupancies_list, axis=2)
std_occupancies  = np.std(bs_occupancies_list, axis=2)

fig, ax = plt.subplots(1,2, figsize=(10,4))
fig.tight_layout(pad=4)

ax[0].errorbar(L_list, mean_occupancies[:,0], yerr=std_occupancies[:,0], capsize=5)
ax[0].errorbar(L_list, mean_occupancies[:,1], yerr=std_occupancies[:,1], capsize=5)
ax[0].set_xlabel('number of binding sites')
ax[0].set_ylabel('mean occupancy')
ax[0].legend(['$J=0$','$J=2.6$'])

ax[1].plot(L_list, std_occupancies, '-.')
ax[1].set_xlabel('number of binding sites')
ax[1].set_ylabel('standard deviation of occupancy')
ax[1].legend(['$J=0$','$J=2.6$'])
```

For very low value of htf, the mean and standard deviation of the occupancy are barely affected by clustering, since the probability of having two TFs next to each other is so low that it almost never happens and thus clustering can not occur.

We can achieve an occupation close to 1 at the binding sites despite having a low htf by increasing the binding energy.
When eb is made too high the occupation in the absense of clustering also goes towards 1, and clustering is not much of a benefit anymore. There is likely an optimal value of eb (for htf = -5 this seems to be around eb = 5 or 6). 
Perhaps longer run times are required to see the effect of the clustering since htf is so low.

# Experiment 3: Binding site occupancy for 6 binding sites with padding

## Experiment 3.1: Average binding site occupancy against $h_{tf}$

```python
from Experiments import average_occupancy_plot

htf_list = np.linspace(-17,-13,9)
eb = 10.5
T = 22500000
runs = 20

L = 6
padding = 9
sites = np.arange(padding,L+padding)

fig, ax = plt.subplots(1,2, figsize=(12,5))
fig.tight_layout(pad=3)

# Without clustering
J = 0
occupancies_list1, _, _, _ = average_occupancy_plot(runs, htf_list, (eb,T), 'htf', '$h_{tf}$', TF_Grid_1D, 
                                                    NSites=len(sites), L=L+2*padding, J=J, sites=sites, ax=ax[0])
ax[0].set_xlim([htf_list[0],htf_list[-1]])
ax[0].set_ylim([0,1])
ax[0].set_title(f'J = {J}')

# With clustering
J = 4.5
occupancies_list2, _, _, _ = average_occupancy_plot(runs, htf_list, (eb,T), 'htf', '$h_{tf}$', TF_Grid_1D, \
                                                    NSites=len(sites), L=L+2*padding, J=J, sites=sites, ax=ax[1])
ax[1].set_xlim([htf_list[0],htf_list[-1]])
ax[1].set_ylim([0,1])
ax[1].set_title(f'J = {J}')
```

Note that the standard deviation is now 1/sqrt(NSites) times the the standard deviation in the case of a single binding site.

Clustering seems to make the curve steeper, but also seems to increase the variance close to the steepest part of the curve. Could the increased steepness outweight the increased variance?

```python
plt.figure()
plt.plot(htf_list, np.std(occupancies_list1,1))
plt.plot(htf_list, np.std(occupancies_list2,1))
plt.xlabel('$h_{tf}$')
plt.ylabel('standard deviation of average bindign site occupancy')
```

Note that for low htf, the standard deviation is increased in the case of clustering, but for higher htf the standard deviation is actually decreased. For the parameters used above, the boundary between those two regimes seems to be around htf=-5.
