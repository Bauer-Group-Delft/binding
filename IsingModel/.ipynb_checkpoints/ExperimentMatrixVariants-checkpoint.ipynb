```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta  = 1
htf   = -15
eb    = 11
J     = 4
L     = 6
sites = np.arange(L)

htf_list = np.linspace(htf-2,htf+2,41)

T = 2*22500000
steps = 1000

periodic = False
```

Note that with non-periodic BCs, the boudnary sites will never be considered to be inside the cluster, since they always have less neighbors.

# Normal matrix

```python
from MarkovChainClass import create

M = create(L, htf, eb, J, sites, periodic=periodic, mode='standard')
mat = M.create_Matrix(sparse=False)

plt.imshow(mat)
plt.xticks(np.arange(M.Nstates), M.str_states, rotation=90)
plt.yticks(np.arange(M.Nstates), M.str_states)
plt.gca().xaxis.set_ticks_position('top')
plt.colorbar()
```

```python
from MarkovChainClass import solve_MasterEquation, solve_full_matrix_htf

t, v = solve_MasterEquation(M, T, steps)
v_compact1 = M.states.compact_states(v, sites)

_, occupancy1 = solve_full_matrix_htf(L, htf_list, eb, J, sites, T=T, periodic=periodic, mode='standard')

plt.figure()
plt.plot(t, v_compact1)
plt.xlabel('$t$')
plt.xlabel('Number of TFs bound')
```

# Comparison

```python
plt.figure()
plt.plot(t, v_compact1, '-')
plt.gca().set_prop_cycle(None)
plt.plot(t, v_compact3, '--')
plt.gca().set_prop_cycle(None)
plt.plot(t, v_compact4, '-.')
plt.gca().set_prop_cycle(None)
plt.plot(t, v_compact5, ':')
plt.xlabel('$t$')
plt.legend([f'{i} TFs bound' for i in range(len(sites)+1)])
```

At low concentrations (htf = -3, eb = 1, J = 1, L = 4, all binding sites) the model behave similar. The matrix where appearance is only possible next to already present TFs has a slightly higher probability of having 0 or 1 TFs bound, since binding of more than 1 TF is slowed down by the fact that the next TFs can only bind next to the already present one. The matrix where disappearance is prevented at the interior of the cluster has a slightly higher probability to have 6 TFs bound, since clusters are less likely to shrink or disappear.

At high concentrations (htf = -1) these effect become stronger.

Interesting here to note is that preventing TFs from disappearing from inside the cluster allows for obtaining high occupancies at lower cocnentrations. This ight mean that this model can obtain the steep response of the hunchback expression at the right concentration with less extreme valued of $eb$ and $J$, which might also lead to less noise.

```python
plt.plot(htf_list, occupancy1, label='standard')
plt.plot(htf_list, occupancy2, label='only appearance')
plt.plot(htf_list, occupancy3, label='nucleation and growth')
plt.plot(htf_list, occupancy4, label='disappearance through shrinking')
plt.plot(htf_list, occupancy5, label='faster appearance in cluster')
plt.legend()
```

For an averaging time of 10 minutes, there is not much difference between the standard model and the disappearance thorugh shrinking model, since the clusters are still assembling and there has not been enough time for a significant amount of molecules to fall off again. When running for (100x) longer however, there is a clear difference.

The faster appearance in cluster model model has higher occupancy than the disappearance through shrinking model. Likely because it speeds up merging of clusters, thereby stabilizing them. This might not be entirely realistic, since this can already happen for single molecules that are 1 site apart, without taking into account the time required to build up a cluster around these points. Therefore most realistic occupancy is expected to lie in between the results obtained for the disappearance through shrinking and faster appearance in cluster models.
