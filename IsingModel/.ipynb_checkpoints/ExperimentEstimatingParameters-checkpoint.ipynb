# $h_{tf}$ range from exponential concentration gradient

```python
R            = 3                   # radius nucleus in um
VNucleus     = 4/3*np.pi*R**3      # um^3
LSite        = 4                   # nm, 4 nm is around 12 bp
VSite        = (LSite*10**-3)**3   # um^3
NSitesTot    = VNucleus/VSite

x_rel_threshold = 0.48             # (Gregor et al., 2007)
c0              = 33               # molecules / um^3 = 0.6*cmax =~ cmax*(10**-9)*NAvogadro/(10**15) 
length_const    = 0.25             # =~ -x_rel_threshold/np.log(cswitch/c0)

beta     = 1
c        = lambda x   : c0*np.exp(-x/length_const)
htf      = lambda c   : 1/beta*np.log(c/(1-c))                    # Note that c must be in molecules/grid cell
fillFrac = lambda htf : np.exp(beta*htf)/(np.exp(beta*htf)+1)

x_rel              = np.linspace(0,1,101)     # relative position along the embryo
molecules_per_nucl = c(x_rel)*VNucleus
molecules_per_site = c(x_rel)*VSite 
htf_along_embryo   = htf(molecules_per_site)
htf_at_threshold   = htf(c(x_rel_threshold)*VSite)

fig, ax = plt.subplots(1,3, figsize=(10,3))
fig.tight_layout(pad=3.0)
ax[0].plot(x_rel, molecules_per_nucl)
ax[0].set_xlabel('relative position, x')
ax[0].set_ylabel('Number of molecules per nucleus, N')
#ax[0].set_title('Number of molecules per nucleus')
ax[1].plot(x_rel, molecules_per_site, label='concentration')
ax[1].plot(x_rel, fillFrac(htf_along_embryo), ':', label='fill fraction')
ax[1].set_xlabel('relative position, x')
ax[1].set_ylabel('Number of molecules per site, N')
ax[1].legend()
#ax[1].set_title('Number of molecules per site (filling fraction)')
ax[2].plot(x_rel, htf_along_embryo, label='$h_{tf}$')
ax[2].scatter(x_rel_threshold, htf_at_threshold, label='switching point')
ax[2].set_xlabel('relative position, x')
ax[2].set_ylabel('$h_{tf}$')
ax[2].legend()
#ax[2].set_title('$h_{tf}$')

print(f'Total number of sites in a nucleus: {int(NSitesTot)}')
print(f'Number of molecules per nucleus: {c(0)*VNucleus} (anterior) to {c(1)*VNucleus} (posterior).')
print(f'Relevant htf range: {htf(c(0)*VSite)} to {htf(c(1)*VSite)}.')
print(f'The boundary of Hunchback expression occurs around relative position {x_rel_threshold}.\nAt this position, the concentration is {c(x_rel_threshold)} molecules/um^3, corresponding to an htf of {htf_at_threshold}.')
```

 For grid cells with a size of 4 nm based on the distance between binding sites on the synthetic enhancer, we obtain a realistic range of htf values between -13 and -17. The switching point should be around $h_{tf}=-15$.

# $e_b$ from retention times

```python
D               = 0.3                     # um^2/s Diffusion constant fo Bicoid (Oliver Grimm and Wieschaus, 2010)
grid_cell_size  = 4                       # nm
time_conversion = (grid_cell_size*10**-3)**2/(2*D) # MSD = 2Dt (1D diffusion through slice)

RT_long = 1 #second, long retention time from (Mir, 2017)
RT_short = 0.1 #second = 100 miliseconds, short retention time from (Mir, 2017)
print(f'Dwell time at a non-binding site: {time_conversion} seconds.')

print('Method 1: Assume long retention time is specific binding without clustering')
k_off = 1/(RT_long/time_conversion) # k_{off} = e^{-beta*eb}
eb = -(1/beta)*np.log(k_off)
print(f'- Binding energy estimated from a retention time of {RT_long} second: {eb}.')

print('Method 2: assume short retention time is clustering and long retention time is binding with clustering.')
k_off = 1/(RT_short/time_conversion) # k_{off} = e^{-beta*eb}
n = 2
J = -(1/beta)*np.log(k_off)/n
print(f'- Clustering strength estimated from a retention time of {RT_short} second assuming {n} neighbors: {J}.')
k_off = 1/(RT_long/time_conversion) # k_{off} = e^{-beta*eb}
eb = -(1/beta)*np.log(k_off)-n*J
print(f'- Binding energy estimated from retention time ratio {RT_long/RT_short}: {eb}.')

print('Method 3: assume short retention time is binding without clustering and long retention time is binding with clustering.')
k_off = 1/(RT_short/time_conversion) # k_{off} = e^{-beta*eb}
eb = -(1/beta)*np.log(k_off)
print(f'- Binding energy estimated from a retention time of {RT_short} second: {eb}.')
k_off = 1/(RT_long/time_conversion) # k_{off} = e^{-beta*eb}
n = 2
J = (-(1/beta)*np.log(k_off)-eb)/n
print(f'- Clustering strength estimated from retention time ratio {RT_long/RT_short}: {J}.')
```

# Fitting Hill equation to find $J$

The switching point or halfway piont occurst at a relative position x/L=0.48 along the embryo, where the $h_{tf}$ value is approximately -15. 

According to (Gregor, 2007) the Hill coefficient $n$ is around 5.

## Optimal $e_b$ and $J$ for $n=5$

*Mean square error:*

```python
from MarkovChain import solve_steady_state_htf

def calculate_MSE(beta, htf_05, htf_list, eb_list, J_list, L, sites, T = None):
    htf_to_c   = lambda htf : 1/(1+np.exp(-beta*htf))
    c_to_htf   = lambda c : 1/beta*(np.log(c)-np.log(1-c))

    n          = 5
    c_05       = htf_to_c(htf_05)
    conc_range = htf_to_c(htf_list)
    Hill_eqn   = lambda c, c_05, n : 1/(1+(c_05/c)**n)         # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)
    hill_curve = Hill_eqn(conc_range, c_05, n)

    MSE = np.zeros((len(eb_list),len(J_list))) 
    for i,eb in enumerate(eb_list):
        for j,J in enumerate(J_list):
            print(f'eb = {eb}, J = {J}.')

            if T == None:
                _, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites)
            else:
                _, occupancy = solve_full_matrix_htf(beta, htf_list, eb, J, L, sites, T=T)

            MSE[i][j] = np.mean((occupancy-hill_curve)**2)
            clear_output(wait=True)

    (eb_opt, J_opt) = np.unravel_index(MSE.argmin(), MSE.shape)

    plt.figure()
    im = plt.imshow(MSE)
    plt.scatter(J_opt, eb_opt)
    plt.xlabel('$J$')
    plt.ylabel('$e_b$')
    plt.colorbar(im)    
    plt.title('MSE')

    return eb_opt, J_opt
```

```python
L          = 6
sites      = np.arange(L)

beta       = 1
htf_05     = -15
htf_list   = np.linspace(htf_05-2,htf_05+2,41)
eb_list    = np.linspace(0,15,16)
J_list     = np.linspace(0,15,16)

T          = 22500000

eb_opt1, J_opt1 = calculate_MSE(beta, htf_05, htf_list, eb_list, J_list, L, sites)
print('Best fit (steady state):', (eb_opt1, J_opt1))
eb_opt2, J_opt2 = calculate_MSE(beta, htf_05, htf_list, eb_list, J_list, L, sites, T=T)
print(f'Best fit (T={T}):', (eb_opt2, J_opt2))
```

```python
from MarkovChain import create_Matrix, solve_MasterEquation, compact_v
from ipywidgets import interactive


def simulation(htf_list, J, eb):
    beta       = 1
    L          = 6
    sites      = np.arange(L)
    T          = 22500000
    
    _, occupancy = solve_full_matrix_htf(beta, htf_list, eb, J, L, sites, T=T)
    
    return occupancy

## -------------------------------- Define parameters --------------------------------

beta       = 1
htf_to_c   = lambda htf  : 1/(1+np.exp(-beta*htf))
c_to_htf   = lambda c    : 1/beta*(np.log(c)-np.log(1-c))

htf_05     = -15
htf_list   = np.linspace(-14,-16,101)
conc_range = htf_to_c(htf_list)

c_05       = htf_to_c(htf_05)
Hill_eqn   = lambda c, n : 1/(1+(c_05/c)**n)                # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)


## -------------------------------- Fit to hill curve --------------------------------

(J_opt, eb_opt), cov = sp.optimize.curve_fit(simulation, htf_list, Hill_eqn(conc_range, 5), bounds = ([0,0],[-htf_05, -htf_05]))
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

print('Optimal J:', J_opt)
print('Optimal e_b:', eb_opt)
plt.figure()
plt.plot(htf_list, simulation(htf_list, J_opt, eb_opt), label=f'simulation result, J = {J_opt}, eb = {eb_opt}')
plt.plot(htf_list, Hill_eqn(conc_range, 5), label='Hill equation')
plt.xlabel('$h_{tf}$')
plt.ylabel('occupancy')
plt.legend()
```

# Ignoring transient in average

```python
from Predictions import predict_occupancy
from Tests import time_mean

def calculate_occupancyAfterTransient(datafile, transient=3*22500000):
    t, NBoundPerSite = datafile.get_NBoundPerSite()
    
    # Remove transient
    idx = np.searchsorted(t, transient, side='right')
    t = np.concatenate(([transient], t[idx:]))
    NBoundPerSite = NBoundPerSite[idx-1:]
    
    mean_occupancy = time_mean(t, NBoundPerSite)
    return mean_occupancy

def repeat_runs(Simulation, function, datafile, runs, T, max_timesteps=int(1e6)):
    outputs = []
    
    for run in range(runs):
        print('run',run)
        Simulation.reinitialize()
        Simulation.run(T, max_timesteps)
        
        # Store results
        outputs.append(function(datafile))
        clear_output(wait=True)
        
    return outputs

def order_parameters(x, ys, mode):    
    if mode == 'htf':
        return x, ys[0], ys[1]
    
    if mode == 'eb':
        return ys[0], x, ys[1]
    
    if mode == 'T':
        return ys[0], ys[1], x
    
    print('Error: Mode not recognised:', mode)
    raise ValueError

def calculate_occupancies_htf_eb_T(runs, xs, ys, mode, grid_class, **kwargs):
    htfs, ebs, Ts = order_parameters(xs, np.repeat(np.transpose([ys]),len(xs),axis=1), mode)
    
    datafile = DataFile(None)
    
    if mode == 'T':
        SimulationGrid = grid_class(htf=ys[0], eb=ys[1], **kwargs)
        Simulation = GillespieSim(SimulationGrid, datafile)
    
    occupancies_list = []
    for htf, eb, T in zip(htfs, ebs, Ts):
        #print(f'{mode}:',x)
        if mode in ['htf', 'eb']:
            SimulationGrid = grid_class(htf=htf, eb=eb, **kwargs)
            Simulation = GillespieSim(SimulationGrid, datafile)
            # Note that we do not have to regenerate the datafile, since it is reinitialized at the beginning of 
            # every run in repeat_runs.

        occupancies = repeat_runs(Simulation, calculate_occupancyAfterTransient, datafile, runs, T)
        occupancies_list.append(occupancies)
    
    return occupancies_list
        
def calculate_expectations(xs, ys, mode, NSites=1):        
    ps = []
    sigmas = []
        
    for x in xs:
        htf, eb, T = order_parameters(x, ys, mode) # This is probably not optimal in terms of computational power required.
        p, sigma = predict_occupancy(htf, eb, T, NSites) 
        ps.append(p)
        sigmas.append(sigma)

    return np.array(ps), np.array(sigmas)

def plot_occupancy(xs_list, ys_list, label_list, xlabel, ylabel, x=None, y=None, sigma=None, mode=2, ax=None):
    if ax == None:
        fig, ax = plt.subplots()
    
    if not (np.any(x==None) or np.any(y==None) or np.any(sigma==None)):
        # Plot expectation
        ax.plot(x, y, c='k', linestyle='-.', label='predicted mean')
        ax.plot(x, y + sigma, c='grey', linestyle=':', label='predicted standard deviation')
        ax.plot(x, y - sigma, c='grey', linestyle=':')
        ax.fill_between(x, y - sigma, y + sigma, color='whitesmoke')
    
    # Plot mean and stdev of occupancy over runs against varied parameter
    # (occupancy = average binding over time for 1 run)
    for xs, ys, label in zip(xs_list, ys_list, label_list):
        ax.scatter(xs, ys, label=label)
        
    if mode == 1:
        ax.errorbar(np.mean(xs_list,1), np.mean(ys_list,1),  xerr=np.std(xs_list,1), yerr=np.std(ys_list,1), linestyle='', c='red', capsize=5, label='standard deviation')
    elif mode == 2:
        ax.fill_between(np.mean(xs_list,1), np.mean(ys_list,1)-np.std(ys_list,1),  np.mean(ys_list,1)+np.std(ys_list,1), label='standard deviation', color='gray', alpha=0.3)
    else:
        print('Warning: Unknown mode.')
        
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()

def average_occupancy_plot_afterTransient(runs, xs, ys, mode, symbol, grid_class, NSites=1, predict=True, plot_mode=2, ax=None, **kwargs):
    
    # Calculate occupancies
    occupancies_list = calculate_occupancies_htf_eb_T(runs, xs, ys, mode, grid_class, **kwargs)
    # **kwargs are the extra arguments required for the given grid_class.
    
    labels = [f'{symbol}$ = {x}$' for x in xs]
    
    if predict:
        # Calculate expectations    
        x_arr = np.linspace(np.min(xs),np.max(xs),1000)
        ps, sigmas = calculate_expectations(x_arr, ys, mode, NSites)
        # Plot mean and stdev of occupancy against htf
        plot_occupancy(np.matlib.repmat(xs, runs, 1).T, occupancies_list, labels, symbol, 'mean occupancy', x_arr, ps, sigmas, ax=ax, mode=plot_mode)
        return occupancies_list, x_arr, ps, sigmas
   
    plot_occupancy(np.matlib.repmat(xs, runs, 1).T, occupancies_list, labels, symbol, 'mean occupancy', mode=plot_mode, ax=ax)
    return occupancies_list
```

```python
## -------------------------------- Define parameters --------------------------------

eb    = 11.5
J     = 3.9
L     = 26
sites = np.arange(6)+10
T     = 4*22500000
runs  = 20

LSite             = 4                   # nm
VSite             = (LSite*10**-3)**3   # um^3 (= 4nm x 4nm x 4nm) 4 nm is around 12 bp
c0                = 33                  # molecules / um^3 = 0.6*cmax =~ cmax*(10**-9)*NAvogadro/(10**15) 
length_const      = 0.25                # =~ -x_rel_threshold/np.log(cswitch/c0)
x_rel_threshold   = 0.48                # (Gregor et al., 2007)

beta     = 1
c        = lambda x   : c0*np.exp(-x/length_const)
htf_to_c = lambda htf : 1/(1+np.exp(-beta*htf))
c_to_htf = lambda c   : 1/beta*(np.log(c)-np.log(1-c))

positions  = np.linspace(0,1,9)
conc_range = c(positions)
htf_list   = c_to_htf(conc_range*VSite)

occupancies = average_occupancy_plot_afterTransient(runs, htf_list, [eb,T], 'htf', '$h_{tf}$', TF_Grid_1D, \
                                                    NSites=len(sites), L=L, J=J, sites=sites, predict=False)

plt.scatter(htf_list, np.mean(occupancies, axis=1), 100, c='k', label='mean', marker='*')

plt.xlabel('relative position, x/L')
plt.xlim([htf_list[0],htf_list[-1]])
plt.xticks(htf_list, positions)
plt.ylim([-0.01,1.01])
plt.yticks(np.arange(6)*0.2)
plt.title('$e_b$ = '+f'{eb}'+', $J$ = '+f'{J}')

beta       = 1
c_05       = c(x_rel_threshold)
Hill_eqn   = lambda c, n : 1/(1+(c_05/c)**n)                # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)
positions  = np.linspace(0,1,101)
conc_range = c(positions)
htf_list   = c_to_htf(conc_range*VSite)
#plt.plot(htf_list, conc_range/c0, c='darkgreen', label='Exponential gradient', linewidth=3)
plt.plot(htf_list, Hill_eqn(conc_range, 5), c='deeppink', label='Hill equation', linewidth=3, zorder=0)
plt.legend('',frameon=False)
```
