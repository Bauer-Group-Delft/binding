# Experiment 1: Timetrace of distribution of number of bound TFs

Use a Master equation to predict the probability of a number of TFs being bound over time.

```python
from MarkovChain import solve_matrix

## -------------------------------- Define parameters --------------------------------

beta = 1
htf  = -5
eb   = 5
J    = 6

L   = 3
sites = np.arange(L)

T     = 1000
steps = 1000

## -------------------------------- Calculate solution --------------------------------

# Full matrix
ts, states, v, v_compact = solve_matrix(beta, L, htf, eb, J, T, steps, sites=sites)

# Simplified matrix
_, s_states, s_v = solve_matrix(beta, L, htf, eb, J, T, steps, sites=sites, matrix='simplified')

## -------------------------------- Plot solution --------------------------------

fig, ax = plt.subplots(1,2,figsize=(12,4))

# Plot probabilities of states
ax[0].plot(ts, v)
ax[0].legend(states)
ax[0].set_xlabel('time, $t$')
ax[0].set_ylabel('probability')
ax[0].set_title('Probabilities of states over time')

# Plot probabilities of number of TFs bound
ax[1].plot(ts, v_compact)
ax[1].set_prop_cycle(None)
ax[1].plot(ts, s_v, ':')
ax[1].legend(np.arange(L+1))
ax[1].set_xlabel('time, $t$')
ax[1].set_ylabel('probability')
ax[1].set_title('Probabilities of number of TFs bound over time')
```

Note that for htf = -7, eb = 0.9, J = 6 and L = 6 (all binding sites), it takes at least 250000 time units before the system has equilibrated.

When all sites have the same eb and J=0, the solution of the simplified matrix corresponds to that of the full matrix.
If all sites have the same eb and $J>0 but L<=3, all sites with a TF bound are identical, including identical number of neighbors bound, such that a similar simplification can be made as in the simplified model, but with different k_off for the different number of TFs bound (based on the number of neighbors each TF has).

### Finding the slution using eigenvalues and eigenvectors of the simplified matrix

```python
from MarkovChain import create_analyticalMatrix, create_analyticalSimplifiedMatrix, solve_analytically

import sympy as sym
import matplotlib.pyplot as plt

beta = 1
htf  = -3 #-7
eb   = 4  #0.9
L = 2

## Simplified matrix
kon, koff = sym.symbols('k_{on}, k_{off}')
kon_val  = np.exp(beta*htf)
koff_val = np.exp(-beta*eb)

seq = L-np.arange(L)
mat = create_analyticalSimplifiedMatrix(L, seq*kon, np.flip(seq)*koff)
v0 = np.zeros(L+1)
v0[0]=1
sol = solve_analytically(mat, v0, [kon, koff], [kon_val, koff_val])

## Solution plot
T     = 200
steps = 1000
ts = np.linspace(0,T,steps+1)

plt.figure()
plt.plot(ts, [sol(t) for t in ts])
plt.legend(np.arange(L+1))
plt.xlabel('time, $t$')
plt.ylabel('probability')
plt.title('Probabilities of number of TFs bound over time')
```

For the siplified matrix the solutions are linear combinations of exponentials, with rates that are integer multiples of -(k_off+k_on) where the integers are in the range [0, number of sites]. Thus if we can find the eigenvalues for the full matrix, we have an indication of the relevant time scales. However, doing this analytically becomes dificult for matrices representing systems with more than 3 sites (i.e. when the solution to the full matrix no longer corresponds to that of the simplified matrix for J=0 and all sites bidnign sites). It would likely be possible to calculte the values of the eigenvalues by substituting the parameter values, but finding the expressions might be too difficult.

However, all systems seem to have one eigenvalue equal to 0, which represents the stable state reached after equilibration.

For the simplified matrix, the slowest nonzero rate is $1/(k_{on}+k_{off})$. This is an accurate approximation for models where J=0 and all sites have the same binding energy $e_b$. For models with 3 sites or less, the simplified matrix is also accurate when $J>0$, if different $k_{off}$ rates are used for different states, however, this will lead to much more complex rates.

We could numerically calculate the rates for a variety of parameter values and possibly find parameter values for which the equilibration takes longest. Or even calculate the equilibration times for those parameters we intend to run simulations with.

For very large grids with few binding sites, we can possibly estimate the equilibration time by ignoring the binding sites, and calculating the expected equilibration time for such a grid without binding sites. We can then use the simplified matrix and the estimate $1/(k_{on}+k_{off})$ for the slowest rate.

### Comparison of matrix solution with eigenvalues

```python
from MarkovChain import solve_matrix

## -------------------------------- Define parameters --------------------------------

beta = 1
htf  = -3
eb   = 0.9
L   = 3

# For full matrix:
# J    = 0
# sites = np.arange(L)

T     = 50
steps = 1000


## -------------------------------- Calculate solution --------------------------------

# Full matrix
ts, states, v = solve_matrix(beta, L, htf, eb, J, T, steps, sites=sites, matrix='simplified')

# Eigenvalues for the simplified matrix
kon_expr  = np.exp(beta*htf)
koff_expr = np.exp(-beta*eb)
eigenvalues = np.arange(L+1)*(kon_expr+koff_expr)
print(eigenvalues)

# Rescale v
v_starting_at_1 = v[0]*v+(1-v[0])*(1-v)
v_rescaled = np.abs((v - v[-1])/(v[0] - v[-1])) # relative distance to stady state


## -------------------------------- Plot comparison --------------------------------

fig, ax = plt.subplots(1,2,figsize=(12,4))

ax[0].semilogy(ts, v_starting_at_1)
for k in eigenvalues:
    ax[0].semilogy(ts, np.exp(-k*ts), ':', c='r')
ax[0].legend(states)
ax[0].set_ylim([1e-1,10])
ax[0].set_title('Time evolution starting from 1')

ax[1].semilogy(ts, v_rescaled)
for k in eigenvalues:
    ax[1].semilogy(ts, np.exp(-k*ts), ':', c='r')
ax[1].legend(states)
ax[1].set_ylim([1e-10,10])
ax[1].set_title('Rescaled time evolution')
```

Plotting as semilogy plot does not give a single straight line with the slope representing the rate because the solutions are sums of exponentials. However, for regions where one of the exponentials dominates, the lines are approximately straight and their slopes are equal to the rates.

Not that due to the rescaling there are places where the solution crosses zero (which in the original solution is the steady state value). At those locations peaks pointing towards -infinity will be observed. This also occurs at the end of the graph, since we subtract the final value in the rescaling.

We observe that the exponentials with higher rates decay very fast and only play a role in the initial part of the solution. The smallest non-zero rate is dominant for a reasonable amount of time before that also decays and the solution becomes constant.
Note that the rescaled solution seems to take a lot longer before the lowest nonzero rate dies out than the original solution. This is because by subtracting the steady state, we essentially remove the term with rate 0.

### Estimating equilibration time based on eigenvalues of the full matrix

```python
from MarkovChain import create_Matrix

import math as m

beta = 1
htf  = -1
eb   = 1
J    = 2.6

L   = 6
sites = np.arange(L)

## -------------------------------- Estimate equilibration time --------------------------------

# Using normal matrix
_, mat, _ = create_Matrix(beta, L, htf, eb, J, sites)
print('Matrix created. Now solving for eigenvalues...')
eigenvalues, eigenvectors = np.linalg.eig(mat)

print(np.sort(-eigenvalues), '\n')
rates = np.sort(np.abs(-eigenvalues))
print('The estimated time constant is:', 1/rates[1])
print(f'It is adviced to run for several times this time, e.g. T={m.ceil(5/rates[1])}.')
```

### Eigenvalues as a function of $h_{tf}$

```python
from MarkovChain import create_Matrix

beta = 1
htf_list = np.linspace(-10,1,100)
eb   = 0
J_list = [0,1,3]

L = 4
sites = np.arange(L)

## -------------------------------- Calculate eigenvalues --------------------------------

kon_expr  = lambda htf : np.exp(beta*htf)
koff_expr = np.exp(-beta*eb)
eigs = np.array([i*(kon_expr(htf_list)+koff_expr) for i in np.arange(L+1)])

fig, axs = plt.subplots(1,len(J_list), figsize=(12,5))

for i, J in enumerate(J_list):
    rates = []

    _, mat, _ = create_Matrix(beta, L, htf_list[0], eb, J, sites)
    lower_lim = -mat.diagonal()
    
    for htf in htf_list:
        _, mat, _ = create_Matrix(beta, L, htf, eb, J, sites)
        eigenvalues, eigenvectors = np.linalg.eig(mat)
        rates.append(np.sort(-eigenvalues))

    axs[i].plot(htf_list, rates)
    axs[i].plot(htf_list, eigs.T, ':', c='k')
    axs[i].scatter([htf_list[0]]*len(lower_lim), lower_lim, marker='x')
    axs[i].plot(htf_list, np.exp(beta*htf_list)+np.exp(-beta*(eb+2*J)), 'r--')
    axs[i].scatter(-eb-J,0,marker='o')
    axs[i].set_xlabel('$h_{tf}$')
    axs[i].set_ylabel('eigenvalues')
    axs[i].set_title(f'J={J}')
    #axs[i].set_ylim([0,1])
```

Simplified matrix: The eigenvalues have the form $-n(k_{on} + k_{off})$. This means that the eigenvalues are always equally spaced. When $h_{tf}$ is very negative, $k_{on}=e^{\beta h_{tf}}$ is very small and the eigenvalues will approach $-nk_{off}$, When $h_{tf}$ is very large, $k_{on}$ is large and the eigenvalues approach $-nk_{on}$, which keeps increasing exponentially with $h_{tf}$.

Full matrix: Note that when the on and off rates are the same for all sites (i.e. same $e_b$ and no clustering), some eigenvalues should have a multiplicity higher than 1 and the number of different eigenvalues should be L+1 as for the simplified matrix.

It seems like for high htf, the clustering strength does not affect the eigenvalues much and the eigevalues are the same as for the simplified matrix, but for low htf, a higher J causes the eigenvalues to converge to the value of a lower eigenvalue. Though there are some lines that seem to converte to a value between 0 and the first nonzero eigenvalue. The lines seem to converge to a lower value as J increases.

For all $h_{tf}$, the eigenvalues of the full matrix seem to be smaller or equal than those of the simplified matrix. This suggests that equilibration of system described by the full matrix takes longer than of that described by the simplified matrix. This effect is stronger when $J$ is larger and $h_{tf}$ is smaller. Bigger $J$ means lower $k_{off}$ (when clustering). This would indeed be expected to slow down equilibration. The effect of k_{off} seems to be more improtant when $h_{tf}$ and thus $k_{on}$ is small. This is the case for the simplified matrix, and now seems to also be true for the full matrix.
```python
from MarkovChain import create_Matrix

beta = 1
htf_list = [-3,-7]
eb   = 0.9
J_list = np.linspace(0,6,100)

L = 3
sites = np.arange(L)

## -------------------------------- Calculate eigenvalues --------------------------------

kon_expr  = lambda htf : np.exp(beta*htf)
koff_expr = np.exp(-beta*eb)

fig, axs = plt.subplots(1,len(htf_list), figsize=(12,5))

for i, htf in enumerate(htf_list):
    rates = []

    for J in J_list:
        _, mat, _ = create_Matrix(beta, L, htf, eb, J, sites)
        eigenvalues, eigenvectors = np.linalg.eig(mat)
        rates.append(np.sort(-eigenvalues))

    eigs = np.array([i*(kon_expr(htf)+koff_expr) for i in np.arange(L+1)])
        
    axs[i].plot(J_list, rates)
    axs[i].plot([J_list[0], J_list[-1]], [eigs, eigs], ':', c='gray')
    axs[i].set_xlabel('$J$')
    axs[i].set_ylabel('eigenvalues')
    axs[i].set_title('$h_{tf}='+f'{htf}$')
    
```

# Experiment 4: Switching rates

```python
N = 1
beta = 1
htf  = -2
eb   = 0

k_on  = np.exp(beta*htf)
k_off = np.exp(-beta*eb)
print('tau_on:', 1/k_on,'tau_off:', 1/k_off)
```

## Expected transition time from N to 0

```python
tau_n = lambda n, tau_next : ((N-n)*k_on)/(n*k_off)*(1/((N-n)*k_on) + 1/((n+1)*k_off) + tau_next)

tau = [0]
for i in range(1,N):
    tau = [tau_n(N-i, tau[0])] + tau

tau_Nto0 = np.sum([tau[n-1]+1/(n*k_off) for n in range(1,N+1)])
print('time from N to 0:', tau_Nto0)
```

*Simplified matrix results:*

```python
from MarkovChain import create_simplifiedMatrix, solve_MasterEquation

L = N
T = 5*tau_Nto0
steps = 2000

# Simplified matrix with absorbing state 0
s_states, s_mat, s_dwell_times = create_simplifiedMatrix(beta, L, htf, eb, sparse=True)
s_mat.data[0,0] = 0 # lower diagonal, first entry
s_mat.data[1,0] = 0 # diagonal, first entry
#print(s_mat.toarray())
v0=np.zeros_like(s_states)
v0[-1]=1
ts, s_v = solve_MasterEquation(s_states, s_mat, T, steps, v0=v0)


## -------------------------------- Plot solution --------------------------------

arival_probability = (s_v[1:,0]-s_v[:-1,0])
average_arival_time_upper  = np.sum(ts[1:]*arival_probability)
average_arival_time_lower  = np.sum(ts[:-1]*arival_probability)
average_arival_time_middle = np.sum((ts[:-1]+ts[1]/2)*arival_probability)
print('calculated time:', average_arival_time_lower, average_arival_time_middle, average_arival_time_upper)

# Plot probabilities of all sites bound
plt.figure()
plt.plot(ts, s_v, ':')
plt.plot(ts[1:], arival_probability/(T/steps))

x = np.linspace(0,T,1000)
plt.plot(x, 1/tau_Nto0*np.exp(-x/tau_Nto0), '--k')

plt.legend(np.concatenate((np.arange(L+1), ['arival time distribution', 'exponential distribution'])))
plt.xlabel('time, $t$')
plt.ylabel('probability')
plt.title('Probabilities of number of TFs bound over time')
```

Note how the rate of change of state N resembles the probability of state N-1 in shape. This makes sense, since the rate of change of state N is equal to the rate from state N-1 to state N multiplied by the probability of state N-1. (Since the reverse rate is set to 0.)

Note that the exponential distribution does not match the actual distribution.

*Full matrix results:*

# Experiment 5: Correlation length at steady state

```python
from MarkovChain import calculate_steady_state, compact_v

beta = 1
htf  = -15
eb   = 10.5
J    = 4.5
L    = 10
sites = np.arange(L)

states, int_states, steady_state = calculate_steady_state(L, beta, htf, eb, J, sites)
idx = int(L/2)

# Calculate correlation
int_states = np.array(int_states)
int_states[np.where(int_states==0)] = -1
mean = np.sum([steady_state[i]*int_states[i] for i in range(len(int_states))], axis=0)

cor_mean = mean*mean[idx] # If J=0, the mean is 0.
correlation_zero_mean = np.sum(np.array([steady_state[i]*(int_states[i]*int_states[i][int(L/2)]) for i in range(len(int_states))]), axis=0)
correlation = correlation_zero_mean - cor_mean

plt.figure()
plt.plot(np.arange(L), correlation, label='Master equation')
x = np.linspace(0,L-1,1000)
xi = -1/np.log(np.tanh(beta*J/4))
plt.plot(x, np.exp(-np.abs(x-int(L/2))/xi), ':k', label = 'Ising model, J/4')
plt.xlabel('Site')
plt.ylabel('Correlation')
plt.ylim([0,1])
plt.legend()
```

Calculating the steady state can likely be made more efficient by building up the states by adding a 0 or 1 to the previous state and calculating the new energy by simply adding the change in energy by the added state to the energy of the previous sate.

This calculation only works for small grids, in which case the periodic BCs might cause artfiacts, by leading to increased corrlations. It is important to think about how the boundaries affect the results, also when including depletion.

*Correlation function at other times:*

```python
from MarkovChain import calculate_steady_state, create_Matrix, calculate_steady_state_correlation_function

def calculate_correlation_function_at_T(mat, t, state_occupancies, p_steady, p_T):
    p_tau = sp.sparse.linalg.expm(mat*t)
    return 1/(len(state_occupancies)**2)*state_occupancies@p_tau@(state_occupancies*p_T)\

beta  = 1
htf   = -3
eb    = 2
J     = 3
L     = 4
sites = np.arange(L)

states, int_states, p_steady = calculate_steady_state(L, beta, htf, eb, J, sites)
state_occupancies = np.sum([int_states[:, i] for i in sites], axis=0)/len(sites)
_, mat, _ = create_Matrix(beta, L, htf, eb, J, sites, sparse=True)

v0 = np.zeros(len(states))
v0[0]=1

plt.figure()
Ts = [0,10,100,1000,1e8]
times = np.linspace(0,100,101)
for T in Ts:
    correlation_function = []
    p_T = sp.sparse.linalg.expm(mat*T)@v0
    for t in times:
        correlation_function.append(calculate_correlation_function_at_T(mat, t, state_occupancies, p_steady, p_T))
    
    plt.plot(times, correlation_function, label=f'T={T}')

correlation_function = []
for t in times:
    correlation_function.append(calculate_steady_state_correlation_function(mat, t, state_occupancies, p_steady))
    
plt.plot(times, correlation_function, ':k', label='steqady state')
    
plt.legend()
plt.xlabel("time, $t$")
plt.ylabel("correlation")
```

If the probabilities of states at the starting point is different, the correlation function might also be different, since states with long dwell times (leading to large correlation times) can be more or less likely.

C(0) is simply the variance, when $T$ is very small, the initial variance is small, but at later times, the variance increases and the mean changes, which could lead to an increase in the correlation.

The correlation function at $T=0$ is always zero, since the occupancy at zero is zero, since it is set by the intitial conditions. The mean at zero is thus also zero, causing the correlation to always be zero, since there is initially no deviation from the mean, so it is not possible to calculate the correlation.
