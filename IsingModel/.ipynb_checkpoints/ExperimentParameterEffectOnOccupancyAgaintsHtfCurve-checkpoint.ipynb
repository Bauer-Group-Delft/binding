# Occupancy against $h_{tf}$


It is important to keep in mind that the averaging time can have a big effect on the results. We start by considering an equilibrated system, after which we will try to determine the effect of the averaging time on these results.

## Part 1: Equilibrated system

First we consider the analytical Master equation for small grids, since this allows us to consider the fully equilibrated system without requireding long rutimes.

### Part 1.1: Equilibrated system with only binding sites

When the system has equilibrated, the halfway point of the curve is at $h_{tf}=h_{tf,0.5}=-(e_b+J)$. However, this is not enough to determine the individual values of $e_b$ and $J$. We wish to know how $e_b$ and $J$ influence the other characteristics of the occupancy against $h_{tf}$ curve, such that we could possibly use those characteristics to determine realistic values of $e_b$ and $J$ from experimental data. 

```python
from MarkovChain import solve_steady_state_htf, plot_cumulative
from ipywidgets import interactive

## -------------------------------- Define parameters --------------------------------

beta  = 1
L     = 4
sites = np.arange(L)

def interactive_plot(eb = 0, J = 0):
    
    htf_05 = -(eb+J)
    htf_list = np.linspace(htf_05-6,htf_05+6,101)

    v_compact, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites)    
    
    fig, ax = plt.subplots()
    plot_cumulative(htf_list-htf_05, v_compact, '$h_{tf}-h_{tf,0.5}$', [f'{i} TFs bound' for i in range(len(sites)+1)], ax=ax)
    ax.plot(htf_list-htf_05, occupancy, linestyle='--', c='k', label='occupancy')
    ax.set_title('$e_b$ = '+f'{eb}'+', $J$ = '+f'{J}')
    ax.legend()
    
w = interactive(interactive_plot, eb=(0.0,10.0,0.05), J=(0.0,10.0,0.05))
display(w)
```

* Increasing $J$ makes the curve steeper, although after a certain value of $J$ the curve does not visibly change anymore.

* Changing $e_b$ does not affect the shape of the curve, which makes sense when looking at the prediction of the mean field approximation: $h_{tf}=\dfrac{1}{\beta}\left[\ln(p)-\ln(1-p)\right]-NpJ-e_b$.
So $e_b$ only causes a shift with respect to the $h_{tf}$ values. It seems to effectively cause a local change in $h_{tf}$ (Although not exactly since $h_{tf}$ affects the on rate, while $e_b$ affects the off-rate. But does that really matter?). Note that these results are for a grid with only binding sites. Changing $e_b$ might have an effect when there are also non-binding sites.

*Relating J to the Hill coefficient $n$:*

```python
from MarkovChain import solve_full_matrix_htf
from ipywidgets import interactive

## -------------------------------- Define parameters --------------------------------

beta  = 1
eb    = 0
L     = 6
sites = np.arange(L)

htf_to_c = lambda htf : 1/(1+np.exp(-beta*htf))
c_to_htf = lambda c : 1/beta*(np.log(c)-np.log(1-c))

n_list = np.linspace(0,10,101)
MSE = np.zeros(len(n_list))

def interactive_plot(J = 4.2):
    
    # Note that different divisions of points along the graph might lead to different results.
    htf_05 = -(eb+J)
    htf_list = np.linspace(htf_05-2,htf_05+2,101)
    conc_range = htf_to_c(htf_list)

    ## -------------------------------- Calculate solution --------------------------------

    v_compact, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites)

    ## -------------------------------- Fit hill curve --------------------------------
    c_05 = htf_to_c(htf_05)
    Hill_eqn = lambda c, n : 1/(1+(c_05/c)**n)         # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)

    for i,n in enumerate(n_list):    
        hill_curve = Hill_eqn(conc_range, n)
        MSE[i] = np.mean((occupancy-hill_curve)**2)

    fig, ax = plt.subplots(2,2,figsize=(8,6))
    fig.delaxes(ax[0][1])
    fig.tight_layout(pad=3)
    
    ax[0][0].plot(n_list,MSE)
    ax[0][0].scatter(n_list[np.argmin(MSE)], np.min(MSE))
    ax[0][0].set_xlabel('n')
    ax[0][0].set_ylabel('mean squared error')
    ax[0][0].set_title(f'J = {J}')
    
    ax[1][0].plot(htf_list, occupancy, label='model results')
    ax[1][0].plot(htf_list, Hill_eqn(conc_range, n_list[np.argmin(MSE)]), label='Hill equation')
    ax[1][0].set_xlabel('$h_{tf}$')
    ax[1][0].set_ylabel('occupancy')
    ax[1][0].set_title(f'n_opt = {n_list[np.argmin(MSE)]}')
    ax[1][0].legend()

    ax[1][1].plot(conc_range, occupancy, label='model results')
    ax[1][1].plot(conc_range, Hill_eqn(conc_range, n_list[np.argmin(MSE)]), label='Hill equation')
    ax[1][1].set_xlabel('concentration, $c$')
    ax[1][1].set_ylabel('occupancy')
    ax[1][1].set_title(f'n_opt = {n_list[np.argmin(MSE)]}')
    ax[1][1].legend()

w = interactive(interactive_plot, J=(0.0,10.0,0.05))
display(w)
```

For a single vlaue of $J$, there is a clear minimum of the mean squared error. Now we wish to determine how this minimum changes with $J$. 

Note that the way we convert htf to concentration means that in the absense of clustering, the concentration and accupancy are equal, i.e. we get a straight line. For $e_b=0$ and $J=0$, the hill equation is thus not a good approximation, since we do not consider a binding reaction, but simply appearing and disappearing. This is a consequence of the assumption we make that $p = c$ in the absense of clustering at non-binding sites. When eb is nonzero, the curve is no longer a line and will have the shape of a hill euqation with $n=1$, since in the absece of clustering binding at the different sites in indepenent.

Note that fixing the halfway point influences the choice for optimal n, since the curve of interest is not precisely a hill curve. Furthermore, different divisions of points along the graph might lead to different results.

Note that while changing $e_b$ does not affect the shape of the occupancy against $h_{tf}$ curve, it does cause a shift. This means that there is a change in shape of the occupancy against concentration curve.

```python
from MarkovChain import solve_steady_state_htf
from ipywidgets import interactive

## -------------------------------- Define parameters --------------------------------

beta  = 1
eb    = 6
L     = 6
sites = np.arange(L)

htf_to_c = lambda htf : 1/(1+np.exp(-beta*htf))
c_to_htf = lambda c : 1/beta*(np.log(c)-np.log(1-c))

J_list = np.linspace(0,10,21)
n_opt = np.zeros(len(J_list))

for j,J in enumerate(J_list):
    htf_05 = -(eb+J)
    htf_list = np.linspace(htf_05-6,htf_05+6,101)
    conc_range = htf_to_c(htf_list)
    
    ## -------------------------------- Calculate solution --------------------------------
    
    v_compact, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites)

    ## -------------------------------- Fit hill curve --------------------------------

    c_05 = htf_to_c(htf_05)
    Hill_eqn = lambda c, n : 1/(1+(c_05/c)**n)         # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)

    n_opt[j], _ = sp.optimize.curve_fit(Hill_eqn, conc_range, occupancy, bounds = ([0,100]))
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

plt.figure()
plt.plot(J_list,n_opt)
plt.xlabel('J')
plt.ylabel('Optimal Hill coefficient, n')
plt.xlim([J_list[0],J_list[-1]])
plt.ylim(bottom=0)
```

$e_b=0$:
- As J increases, the optimal value of $n$ approaches the number of binding sites. 
- At low values of $J$ the curve does not resemble a Hill function and the obtained values for $n$ will be significantly affected by the fixing of the halfway point. The obtained $n$ values for low $J$ are thus not very informative.
- According to (Gregor, 2007) the Hill coefficient $n$ is around 5. This corresponds to a $J$ that is equal to 4.2.

$e_b=15$:
- At J=0, the occupancy against concentration curve is no longer a strainght line and can be fitted with a Hill equation with Hill coefficient 1.

The higher $e_b$, the closer to 1 the occupancy at $J=0$ will be. From $e_b\geq 6$ it is close to 1.

### Part 1.2: Equilibrated system with binding and non-binding sites

```python
from MarkovChain import solve_steady_state_htf, plot_cumulative
from ipywidgets import interactive

## -------------------------------- Define parameters --------------------------------

beta  = 1
L     = 2+2*2
sites = np.arange(2)+2

def interactive_plot(eb = 0, J = 0):
    
    htf_05 = -(eb+J) #-(eb+J/2)
    htf_list = np.linspace(htf_05-6,htf_05+6,101)
    
    ## -------------------------------- Calculate solution --------------------------------

    v_compact, occupancy, v_compact_all, occupancy_all = solve_steady_state_htf(beta, htf_list, eb, J, L, sites, all_sites=True)

    ## -------------------------------- Plot solution --------------------------------

    fig, ax = plt.subplots(1,2,figsize=(12,4))
    
    # Cummultive probabilies
    plot_cumulative(htf_list-htf_05, v_compact_all, '$h_{tf}-h_{tf,0.5}$', [f'{i} TFs bound' for i in range(L+1)], ax=ax[0])
    plot_cumulative(htf_list-htf_05, v_compact, '$h_{tf}-h_{tf,0.5}$', [f'{i} TFs bound' for i in range(len(sites)+1)], ax=ax[1])
    
    # Expected occupancy
    ax[0].plot(htf_list-htf_05, occupancy_all, linestyle='--', c='k', label='occupancy') # expected fraction of binding sites occupied
    ax[1].plot(htf_list-htf_05, occupancy, linestyle='--', c='k', label='occupancy') # expected fraction of binding sites occupied
    
    ax[0].legend()
    ax[1].legend()
    ax[0].set_title('All sites, $e_b$ = '+f'{eb}'+', $J$ = '+f'{J}')
    ax[1].set_title('Binding sites only, $e_b$ = '+f'{eb}'+', $J$ = '+f'{J}')
    
w = interactive(interactive_plot, eb=(0.0,10.0,0.05), J=(0.0,10.0,0.05))
display(w)
```

* Changing $e_b$ in this case indeed affects the occupancy of the full grid. However, as long as there is no clustering, the sites are independent and the occupancy of the binding sites is not affected by $e_b$ (apart form the shift in $h_{tf}$ value).
* When $J>0$, increasing $e_b$ causes a shift to higher $h_{tf}-h_{tf,0.5}$, but also seems to decrease the slope of the graph somewhat. This implies that the shift in $h_{tf}$ is less than the expected $h_{tf,0.5}$. This makes sense since there is now less clustering at the edges of the binding site array, therefore requiring a higher $h_{tf}$ to reach the same occupancy. This effect will likely be smaller for larger binding site arrays, since then the fraction of binding sites with a non-binding site neighbor is less. The shift is caused by the fact that increasing $J$ does not increase the occupancy as much as expected by the mean field approximation, because the non-binding sites have a lower occupancy than the binding sites, while in the mean field approxdimation they are all assumed to be the same. This effect is stronger when $e_b$ is higher, since then there will be a larger difference between occupancy of binding and non-binding sites and thus the occupancy of the non-bindign sites will be lower at the $h_{tf}$ for which the occupancy of the binding sites is 0.5.
* For very high values of $e_b$ we observe that occupancy of the full grid essentially consists of a combination of 2 sigmoidal curves. One for filling up of the binding sites and one for filling up of the non-binding sites. When $e_b$ is very high, the binding sites fill up at a much lower $h_{tf}$ than the non-binding sites and these processes are almost completely separated. This suggests that a comparison between the concentration of TFs at the binding sites and the concentration of TFs in the rest of the cell could be informative of the value of $e_b$, although this might not be as easy to determine from the experimental data. Increasing $J$ leads to less separation, because the occupancy of the non-bidning sites increases.

Thus, the effect of $e_b$ in the occupancy of the binding sites is a boundary effect, which is only observed in the presence of clustering. We now wish to find an expression for the additional shift caused by this effect, such that we can study the effect on the shape of the curve more accurately.

The binding sites at the boundary of the binding site array have only one binding site neighbor. In the limit where $e_b$ is large (compared to $J$?), the non-binding sites will have an occupancy of 0 when the binding sites have an occupancy of 0.5 and the non-binding sites do not contribute at all to the occupancy of the biding sites. The boundary effect is strongest when the binding site array is short. In the limit of large $e_b$, a single binding site will not notice the effect of clustering at all, and we simply have $h_{tf,0.5}=-(e_b)$. For an array of two binding sites, each has a single binding site neighbor and $h_{tf,0.5}=-(e_b+J/2)$. For larger arrays of binding sites, the value of $h_{tf,0.5}$ must be somewhere between $-(e_b+J)$ and $-(e_b+J/2)$. If $e_b$ is smaller, there will still be some contribution of the non-binding sites, especially at high values of $J$, and $h_{tf,0.5}$ will be lower (i.e. more negative).

* In the limit where $e_b>>J$, changing $e_b$ does not seem to affect the shape fo the graph. This makes sense, since a change in shape of the graph indicates changes in cooperativity (e.g. due to clustering). When $e_b$ is large compared to $J$, clustering with non-binding sites does not occur and there is no change in the clustering between binding sites, so there is no reason why the shape of the graph would change. When $e_b$ becomes of the same order of magnitude as $J$ however (right?), clustering with non-binding sites starts having an effect and the graph becomes slightly steeper. (Note that this is technically due to the effect of clustering and not the effect of the binding energy.) 

So the binding energy affects the shape of the curve by changing the position of the halfway point and thereby determining the effect of clustering at this point.

Conclusion: If $J$ is larger, the deviation of the shift from $-(e_b+J)$ could be larger, but the actual deviation depends on the (relative?) value of $e_b$.

*Effect of $e_b$ on shape:*

To see whether $e_b$ affects the shape of the graph in the presence of non-binding sites, we first use the bisection method to determine the shift, and then overlay the curves to compare their shape.

*Effect of $e_b$ on shift:*

```python
from MarkovChain import calculate_steady_state, compact_v, solve_matrix

def bisection(beta, L, htf_init, eb, J, sites, tol = 0.02, periodic=True, T=None):
    # Bisection method
    while (htf_init[1]-htf_init[0]) > tol:

        htf = (htf_init[0]+htf_init[1])/2

        if T==None:
            states, _, v = calculate_steady_state(L, beta, htf, eb, J, sites, periodic)
            v_compact = compact_v(states, np.array([v]), sites)
        else:
            _, states, _, output = solve_matrix(beta, L, htf, eb, J, T, 1, sites=sites)
            v_compact = output[-1]
            
        occupancy = np.sum(v_compact*np.arange(len(sites)+1))/len(sites)   # Expected binding site occupancy

        if occupancy > 0.5:
            htf_init[1] = htf
        else:
            htf_init[0] = htf

    return (htf_init[0]+htf_init[1])/2 # Error = tol/2
```

```python
## -------------------------------- Define parameters --------------------------------

beta  = 1
J_list = [0,1,3,6]
eb_list = np.linspace(0,6,13)

fig, ax = plt.subplots(1,4,figsize=(12,3))
fig.tight_layout(pad=3)

for j,J in enumerate(J_list):
    ax[j].plot(eb_list, -(eb_list+J), ':', c='gray')
    ax[j].plot(eb_list, -(eb_list+J/2), ':', c='gray')
    ax[j].set_xlabel('$e_b$')
    ax[j].set_ylabel('$h_{tf,0.5}$')
    ax[j].set_title(f'J = {J}')

for N in range(2,5):
    L     = N+2*2
    sites = np.arange(N)+2
        
    for j,J in enumerate(J_list):
        htf_05_est = np.zeros(len(eb_list))
        htf_05_est_nonperiodic = np.zeros(len(eb_list))
        for i,eb in enumerate(eb_list):
            print(f'N = {N}, eb = {eb}, J = {J}')
            htf_init = [-(eb+J), -(eb+J/2)]
            htf_05_est[i] = bisection(beta, L, htf_init, eb, J, sites)
            htf_init = [-(eb+J), -(eb+J/2)] # Redefine htf_init, since it is changed in bisection.
            htf_05_est_nonperiodic[i] = bisection(beta, N, htf_init, eb, J, np.arange(N), periodic=False)
            clear_output(wait=True)

        ## -------------------------------- Plot solution --------------------------------
        
        ax[j].plot(eb_list, htf_05_est, color=f'C{N-2}')
        ax[j].plot(eb_list, htf_05_est_nonperiodic, ':', color=f'C{N-2}')
        
```

For a binding site array of 2 binding sites, the shift transitions from $-(e_b+J)$ to $-(e_b+J/2)$ as $e_b$ increases. Thus, if $J$ is larger, the maximum deviation in the shift ($J/2$) is larger. This also means that a larger value of $e_b$ is required to obatain such a large deviation.

For larger binding site arrays, the curves approach some value between $-(e_b+J)$ and $-(e_b+J/2)$ as $e_b$ increases. The larger the binding site array, the lower this value (less deviation from $-(e_b+J)$), but this effect seems to decrease with the number of binding sites (diminishing returns).

As $e_b$ increases, the results approach the limit of a grid of N sites with non-periodic BCs. It seems like this limit can be reasonably approximated by $-(e_b+n_{avg}*J/2)$ were $n_{avg}$ is the average number of neighbors, assuming nonbinding sites have an occupancy of 0, i.e. $n_{avg}=(2*(N-2)+2)/N$.

*Effect of $J$ on shift:*

```python
## -------------------------------- Define parameters --------------------------------

beta  = 1
eb_list = [0,1,3,6]
J_list = np.linspace(0,6,13)

fig, ax = plt.subplots(1,4,figsize=(12,3))
fig.tight_layout(pad=3)

for i,eb in enumerate(eb_list):
    ax[i].plot(J_list, -(eb+J_list), ':', c='gray')
    ax[i].plot(J_list, -(eb+J_list/2), ':', c='gray')
    ax[i].set_xlabel('$J$')
    ax[i].set_ylabel('$h_{tf,0.5}$')
    ax[i].set_title('$e_b$ = '+f'{eb}')

for N in range(2,5):
    L     = N+2*2
    sites = np.arange(N)+2
    for i,eb in enumerate(eb_list):
        htf_05_est = np.zeros(len(J_list))
        htf_05_est_nonperiodic = np.zeros(len(J_list))
        for j,J in enumerate(J_list):
            print(f'N = {N}, eb = {eb}, J = {J}')
            htf_init = [-(eb+J), -(eb+J/2)]
            htf_05_est[j] = bisection(beta, L, htf_init, eb, J, sites)
            htf_init = [-(eb+J), -(eb+J/2)] # Redefine htf_init, since it is changed in bisection.
            htf_05_est_nonperiodic[j] = bisection(beta, N, htf_init, eb, J, np.arange(N), periodic=False)
            clear_output(wait=True)

        ## -------------------------------- Plot solution --------------------------------

        ax[i].plot(J_list, htf_05_est, color=f'C{N-2}')
        ax[i].plot(J_list, htf_05_est_nonperiodic, ':', color=f'C{N-2}')
        
```

If $J$ is higher, clustering becomes more important, also with TFs on nonbinding sites, and the results move away from the results for a grid with only binding sites and nonperiodic BCs (no clustering with sites around the binding site array). When $e_b$ increases, clustering with TFs on nonbinding sites becomes less and less likely leading to the results for the two grids becoming more similar.

*Optimal shift and Hill coefficient for differnt $e_b$ and $J$:*

It would be interesting to plot the shift and optimal n for a grid of $e_b$ and $J$ values. The shift can be determined using the bisection method as above, after which it can be used to fix the shift of the Hill equation and fit $n$. We can then take the absolute difference with the expected shift and Hisll coefficient to determine the most likely combinations of $e_b$ and $J$.

```python
from MarkovChain import solve_steady_state_htf

beta  = 1
L     = 3+2*2
sites = np.arange(3)+2

eb_list = np.linspace(0,15,16)
J_list  = np.linspace(0,15,16)

shifts = np.zeros((len(eb_list),len(J_list))) 
n_opts = np.zeros((len(eb_list),len(J_list))) 

htf_to_c = lambda htf : 1/(1+np.exp(-beta*htf))
c_to_htf = lambda c : 1/beta*(np.log(c)-np.log(1-c))

for i,eb in enumerate(eb_list):
    for j,J in enumerate(J_list):
        print(f'eb = {eb}, J = {J}.')
        
        ## Bisection method
        print('Determining shift...')
        
        htf_init = [-(eb+J), -(eb+J/2)]
        shifts[i,j] = bisection(beta, L, htf_init, eb, J, sites)
        
        ## Fitting Hill equation
        print('Fitting Hill coefficient...')
        
        htf_list = np.linspace(shifts[i,j]-6,shifts[i,j]+6,101)
        v_compact, occupancy = solve_steady_state_htf(beta, htf_list, eb, J, L, sites)

        c_05 = htf_to_c(shifts[i,j])
        Hill_eqn = lambda c, n : 1/(1+(c_05/c)**n)         # https://en.wikipedia.org/wiki/Hill_equation_(biochemistry)
        conc_range = htf_to_c(htf_list)
        n_opts[i,j], _ = sp.optimize.curve_fit(Hill_eqn, conc_range, occupancy, bounds = ([0,10]))
        # https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

        clear_output(wait=True)
    
fig, axs = plt.subplots(1,2, figsize=(6,3))
fig.tight_layout(pad=3)

im1 = axs[0].imshow(shifts)
im2 = axs[1].imshow(n_opts)
plt.colorbar(im1, ax=axs[0])
plt.colorbar(im2, ax=axs[1])

for ax in axs:
    ax.set_xlabel('$J$')
    ax.set_ylabel('$e_b$')
    
axs[0].set_title('Shift')
axs[1].set_title('Hill coefficient')
```

If $e_b=0$, the shift is $-J$, while if $J=0$, the shift is $-e_b$. When both are nonzero, the shift is smaller than $-(eb+J)$. The deviation from the expected shift $-(eb+J)$ seems to be largest when $e_b=J$.

The hill coefficient increases with $J$, but if $e_b$ is higher, a higher $J$ is required to obtain the same steepness.

### Part 2.2: Time averaging

Now that we have a bit of an idea of the effect on the mean occupation, we also would like to know the effect on the standard deviation. 

The standard deviation is affected by the averaging time, the longer one averages (i.e. the larger the averaging time compared to correlation time, $T>>\tau_c$), the smaller the standard deviation. The standard deviation also depends on the relative magnitude of the averaging time and equilibration time. When $T>>\tau_{eq}$ the mean will be barely influenced by the initial conditions and the standard deviation will be lower. Since the equilibration time and correlation times are related, we likely have $T>>\tau_{eq} \Rightarrow T>>\tau_c$.

*Shifting $h_{tf}$ range:*
